```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
data = pd.read_csv('data/2_mul_linreg_homeprices.csv')
data.head()
```

```python
# Data preprocessing
# replace missing bedrooms with median bedroom

median_bedrooms = np.floor(data.bedrooms.median())
```

```python
data.bedrooms = data.bedrooms.fillna(median_bedrooms)
data.head()
```

```python
model = linear_model.LinearRegression()
model.fit(data.drop(columns='price'), data.price)
```

```python
model.predict(np.array([3000, 3, 15]).reshape(1, 3))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
df = pd.read_csv('data/2_1_mul_linreg_hiring.csv')
df.info()
df.head()
```

```python
## data preprocessing
# convert exp to numbers
# fill the missing data

from word2number import w2n
df.experience = df.experience.fillna('zero')
df['experience_num'] = df.experience.apply(w2n.word_to_num)
df.head()
```

```python
df['test_score'] = df.test_score.fillna(np.floor(df.test_score.median()))
df.head()
```

```python
model = linear_model.LinearRegression()
model.fit(df.drop(columns=['experience', 'salary']), df.salary)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict(np.array([10, 10, 12]).reshape(-1,3))
```
