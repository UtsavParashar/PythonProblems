## Naive bayes is based on bayes theoram
* Naive because it makes a naive assumption that the independent features are not corelated
* Bayes because it is based on Bayes Theoram
* Bayes Theoram == P(A|B) = P(B|A)*P(A)/ P(B)
* Bayes Theoram is based on conditional probability = P(A|B) = P(A inteserction B)/P(B)

```python
# data = using titanic dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
# Data Exploration

df = pd.read_csv('data/14_titanic.csv')
df.head()
```

```python
cols = ['PassengerId', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
df = df[cols]
df.head()
```

```python
dummies = pd.get_dummies(df.Sex)
dummies.head()
type(dummies)
```

```python
df = pd.concat([df, dummies], axis=1)
df['female'] = df.female.astype(int)
df['male'] = df.male.astype(int)
df.head()
```

```python
target = df.Survived.values
```

```python
cols = ['PassengerId', 'Pclass', 'Age', 'Fare', 'female', 'male']
df = df[cols]
df['Age'] = df.Age.fillna(df.Age.mean())
df.info()
```

```python
df.groupby('Pclass')[['PassengerId']].count()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, target,)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
