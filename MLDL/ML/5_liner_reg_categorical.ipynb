```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
data = pd.read_csv('data/5_homeprices_onehot.csv')
data
```

```python
one_hot_encoded_towns = pd.get_dummies(data.town)
one_hot_encoded_towns.columns = one_hot_encoded_towns.columns.str.replace(' ', '_')
one_hot_encoded_towns
```

```python
final = pd.concat([data, one_hot_encoded_towns], axis=1)
final.head()
```

```python
# removing a one hot encoded column west_windsor ro avoid multicolinearity or dummy variable trap
X = final.drop(columns=['town', 'west_windsor', 'price'])
y = data['price']
X
```

```python
model = linear_model.LinearRegression()
model.fit(X, y)
```

```python
# Predict price for west windsor 3450 

model.predict(np.array([3450, 0, 0]).reshape(1, -1))
```

```python
# Predict price for monroe township 4400 
model.predict(np.array([4400, 1, 0]).reshape(1, -1))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(X, y)
```
