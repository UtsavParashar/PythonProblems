```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/6_carprices.csv')
data.head()
```

```python
%matplotlib inline

plt.scatter(data.Mileage, data.SellPrice)
```

```python
%matplotlib inline

plt.scatter(data.Age, data.SellPrice)
```

```python
%matplotlib inline

plt.scatter(data.Mileage, data.Age)
```

```python
X = data.drop(columns=['SellPrice']).values
y = data.SellPrice.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test, y_test)
```
