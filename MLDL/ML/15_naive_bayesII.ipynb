```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB, BernoulliNB

```

```python
df = pd.read_csv('data/15_nb_spam.csv')
df.head()
```

```python
df['spam'] = np.where(df.Category == 'ham', 0, 1)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=.25)
```

```python
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()
```

```python
model = BernoulliNB() # or MultinomialNB
```

```python
## Create pipeline of the steps like vectorize, apply model etc so that we don't have to follow these steps manually

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
