```python
def partition(a, start, end):
    pivot = a[start]
    
    while a[start] <= pivot:
        start += 1
        
    while a[end] > pivot:
        end -= 1
    
    a[start],a[end] = a[end], a[start]
    print(a)
    if start >= end:
        pivot, a[end] = a[end], pivot
    
    return end
```

```python
def quick_sort(a, start, end):
    if start < end:
        p = partition(a, start, end)
        quick_sort(a, 0, p-1)
        quick_sort(a, p+1, end)
    return p
```
