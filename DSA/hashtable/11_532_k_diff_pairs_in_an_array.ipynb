```python
# Preferred Solution
# Discuss trade off of the two solution and the write a solution accordingly
# HM - O(n), O(n)
# Two Pointers - O(n log n), O(1)

from typing import List
from collections import Counter


class Solution:
    """
        HASH MAP Solution - Preferred
        Returns the number of unique k-diff pairs in the array.

        A k-diff pair is defined as (a, b) where:
        - |a - b| = k
        - a and b are integers from nums
        - Each unique pair is counted once

        Time Complexity: O(n)
        Space Complexity: O(n)
    """
    def findPairs(self, nums: List[int], k: int) -> int:
        pairs_count = 0
        # Absolute difference cannot be negative
        if k < 0 or not nums:
            return pairs_count
        
        freq_counter = Counter(nums)
        
        if k == 0:
            # Count numbers that appear more than once
            for key, freq in freq_counter.items():
                if freq > 1:
                    pairs_count += 1
        else:
            # Count unique values x such that x + k exists
            for key in freq_counter.keys():
                if key + k in freq_counter:
                    pairs_count += 1
            
        return pairs_count
```

```python
from typing import List

class Solution:
    def findPairs(self, nums: List[int], k: int) -> int:
        """
        Returns the number of unique k-diff pairs using sorting + two pointers.

        Constraints:
        - Each unique pair is counted once
        - Absolute difference equals k

        Time Complexity: O(n log n)
        Space Complexity: O(1) (excluding sorting)
        """

        pairs_count = 0
        if k < 0 or not nums:
            return pairs_count

        nums.sort()
        left = 0
        right = 1
        n = len(nums)

        while left < n and right < n:
            # Ensure pointers are not the same # Eg 1 1 3 4 5
            if left == right:
                right += 1
                continue

            curr_diff = nums[right] - nums[left] 
            if curr_diff == k:
                pairs_count += 1
                left += 1
                right += 1

                # Skip duplicates for left pointer
                while left < n and nums[left] == nums[left-1]:
                    left += 1

                # Skip duplicates for right pointer
                while right < n and nums[right] == nums[right-1]:
                    right += 1
            # Need a larger difference → move right
            elif curr_diff < k:
                right += 1
            # Difference too large → move left
            else:
                left += 1
        
        return pairs_count
```
