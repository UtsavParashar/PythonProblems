#### 430. Flatten a Multilevel Doubly Linked List

* https://leetcode.com/problems/flatten-a-multilevel-doubly-linked-list/description/

```python
# http://chatgpt.com/c/698175a1-3a38-8320-8288-77ef770a988c

"""
# Definition for a Node.
class Node:
    def __init__(self, val, prev, next, child):
        self.val = val
        self.prev = prev
        self.next = next
        self.child = child
"""

class Solution:
    def flatten(self, head: 'Node') -> 'Node':
        """
        Flattens a multilevel doubly linked list into a single-level list
        using DFS (preorder traversal).

        Production considerations:
        - Iterative to avoid recursion depth issues
        - In-place pointer manipulation
        - Explicit nulling of child pointers

        Time Complexity: O(N)
        Space Complexity: O(N) in worst case due to stack
        """
        if not head:
            return head

        stack = []
        current = head

        while current:
            # If current node has a child,
            # we need to process child list before next node
            if current.child:
                # If there's a next node, save it for later
                if current.next:
                    stack.append(current.next)
                    current.next.prev = None  # detach safely

                # Link child as next
                current.next = current.child
                current.child.prev = current
                current.child = None  # required by problem

            # If we reach end of a list and have pending nodes
            if not current.next and stack:
                next_node = stack.pop()
                current.next = next_node
                next_node.prev = current

            current = current.next

        return head
```
