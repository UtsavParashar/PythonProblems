#### 19. Remove Nth Node From End of List

* https://leetcode.com/problems/remove-nth-node-from-end-of-list/description/

#### ['Bloomberg', 'Meta', 'Amazon', 'Apple', 'Uber']

```python
# Single pass solution
# TC, SC - O(n), O(1)
# if you are thinking that in an eg 1->2->3->4->5, where n = 2. what will happen if we do 3.next.next as 4 is also pointing to 5.
# It's fine as nothing is pointing to it so it will be garbage collected or won't fall in the path 1->2->3->5

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeNthFromEnd(self, head, n: int):
        # Edge Cases - n >> len then n % len
        # no nodes - return head
        # single node, n = 1
        # One Pass solution
        # Fast pointer to point the n-1st node
        # Increment fast and slow pointers
        # when fast pointer reaches the end, slow pointer points to the node before the node to be deleted
        # point slow pointer to next.next

        dummy = ListNode(0, head)
        fast = slow = dummy

        # fast pointer to point to nth node
        for _ in range(n):
            fast = fast.next

        # move both fast and slow
        while fast.next:
            fast = fast.next
            slow = slow.next

        slow.next = slow.next.next

        return dummy.next
```

```python
# Two pass solution using len of the linkedlist
# TC - O(n)
# SC - O(1)
# Useful for the edgecase when n could be greater than length n >> l

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeNthFromEnd(self, head, n: int):
        # Two pass solution
        # 1. Find the length of the list
        # 2. Traverse again to reach the node just before the one to remove — that’s the (L - n)th node.
        # 3. Update its next pointer to node.next.next

        dummy = ListNode(0, head)
        length = 1
        curr = head

        while curr.next:
            length += 1
            curr = curr.next

        curr = dummy
        for _ in range(length - n):
            curr = curr.next

        curr.next = curr.next.next
        return dummy.next

        
```

```python
# Credit - https://www.youtube.com/watch?v=fbi4UdayLSA
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def __repr__(self):
        return f'{self.val}, {self.next}'

def remove_nth_node_from_end(head, n):
    dummy = ListNode(0, head) # Dummy node is required for the edge case if the head node is to be removed
    left, right = dummy, head

    # Move right ahead by n steps
    for _ in range(n):
        right = right.next

    # Move both pointers until right reaches the end
    while right:
        left, right = left.next, right.next

    # Remove the nth node from the end
    left.next = left.next.next
    return dummy.next
```

```python
# Create the linked list: 1 -> 2 -> 3 -> 4 -> 5
root = ListNode(1)
root.next = ListNode(2)
root.next.next = ListNode(3)
root.next.next.next = ListNode(4)
root.next.next.next.next = ListNode(5)

# Remove the 2nd node from the end
result = remove_nth_node_from_end(root, 2)

# Print the resulting linked list
result
```
