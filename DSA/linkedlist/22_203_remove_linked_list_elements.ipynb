##### 203. Remove Linked List Elements

* https://leetcode.com/problems/remove-linked-list-elements/description/

##### BBG

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val: int = 0, next: "ListNode | None" = None):
        self.val = val
        self.next = next


class Solution:
    def removeElements(self, head: ListNode | None, val: int) -> ListNode | None:
        """
        Removes all nodes with value == val from a singly linked list.

        Approach:
        Use a sentinel (dummy) node so deletion logic is uniform
        even when the head node must be removed.

        Time Complexity: O(n)
        Space Complexity: O(1)
        """

        sentinel = ListNode(0, head)
        previous = sentinel
        current = head

        while current:
            if current.val == val:
                # skip current node
                previous.next = current.next
            else:
                previous = current

            current = current.next

        return sentinel.next
```

```python
### Follow Up 1 - Remove elements recursively

def removeElements(self, head: ListNode | None, val: int) -> ListNode | None:
    if not head:
        return None

    head.next = self.removeElements(head.next, val)
    return head.next if head.val == val else head
```

```python
### Follow Up 2 - Remove only the first occurence

def removeFirst(head, val):
    sentinel = ListNode(0, head)
    prev, curr = sentinel, head

    while curr:
        if curr.val == val:
            prev.next = curr.next
            break
        prev, curr = curr, curr.next

    return sentinel.next
```
