```python
# https://chatgpt.com/c/695794e5-29c8-8324-84a1-a12a7d95288b
# Stack approach - TC - O(m+n), SC - O(m+n)

from typing import List, Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        # Two approaches
        # Stack approach - TC - O(m+n), SC - O(m+n)
        # Reverse ll approach - TC - O(m+n), SC - O(1)

        # Discuss the tradeoff before choosing the approach
        # In Stack approach lists are not reversed/updated so extra space is required
        # works when the data needs to be kept immutable as same data(ll) might be used
        # by different systems, multiple threads

        # Reverse ll approach is fast and takes less space
        # however, we would be reversing the ll
        # this would affect the mutability of the list
        # even if we reverse the ll again once the job is done then also 
        # there are chances that some other system reads the ll when we reversed it

        # Stack approach -
        # Place the elements of both lists in separate stacks
        # fetch the values from the stack and add them along with carry - LC 2
        # create a node add value to it, point it to head
        # list would be created backwards from node to head

        # Reverse ll approach
        # reverse both lls - LC 206
        # add the elements of ll - LC 2
        # create a node add value to it, point it to head
        # list would be created backwards from node to head

        if not l1:
            return l2
        if not l2:
            return l1

        s1, s2 = [], []
        while l1:
            s1.append(l1.val)
            l1 = l1.next

        while l2:
            s2.append(l2.val)
            l2 = l2.next

        carry = 0
        head = None
        while s1 or s2 or carry:
            v1 = s1.pop() if s1 else 0
            v2 = s2.pop() if s2 else 0

            total = v1 + v2 + carry
            carry = total // 10

            node = ListNode(total%10)
            node.next = head
            head = node

        return head


        
```
