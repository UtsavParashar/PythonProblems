#### 328. Odd Even Linked List
* https://leetcode.com/problems/odd-even-linked-list/description/

#### BBG -IMP

```python
from typing import Optional
import time

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def __repr__(self):
        return f'{self.val}'
class Solution:
    """
        Pointer manipulation

        Reorders the linked list such that all nodes at odd indices
        are grouped first followed by nodes at even indices.

        Indexing is 1-based.

        Time Complexity: O(n)
        Space Complexity: O(1)
        
    """

    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Edge cases: empty list or single node
        if not head or not head.next:
            return head


        odd = head
        even = even_head = head.next

        # Iterate while there are available even and odd nodes ahead
        # iterating on even since even will form the tail
        while even and even.next:
            # Link current odd to the next odd node
            odd.next = even.next
            odd = odd.next
            
            # Link current even to the next even node
            even.next = odd.next
            even = even.next

            print(f'odd - {odd.val} and even - {even.val} ==> {odd.next} {even.next}')
            time.sleep(5)

        # Attach even list at the end of odd list
        odd.next = even_head

        return head

root = ListNode(1)
root.next = ListNode(2)
root.next.next = ListNode(3)
root.next.next.next = ListNode(4)
root.next.next.next.next = ListNode(5)
root.next.next.next.next.next = ListNode(6)
root.next.next.next.next.next.next = ListNode(7)
root.next.next.next.next.next.next.next = ListNode(8)
Solution().oddEvenList(root)
```
