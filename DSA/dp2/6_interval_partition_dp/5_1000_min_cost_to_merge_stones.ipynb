##### 1000. Minimum Cost to Merge Stones

* https://leetcode.com/problems/minimum-cost-to-merge-stones/description/

```python
from typing import List
from functools import lru_cache
from itertools import accumulate

class Solution:
    def mergeStones(self, stones: List[int], k: int) -> int:
        """
            TC - O(n^3/k) # n^2 for interval * n for partition / each iteration in n is k steps
            SC - O(n^2)
        """
        n = len(stones)

        # why (n-1)%(k-1)
        # because n needs to be reduced to 1
        # after each iteration k-1 elements are reduced 
        if (n-1)%(k-1) != 0:
            return -1

        # why [0] + running sum
        dp = [0] + list(accumulate(stones))

        # why dp[r+1] - dp[l]
        def ic(l, r):
            return dp[r+1] - dp[l]
        
        @lru_cache(maxsize=None)
        def compute(l, r):
            if r == l:
                return 0
            
            mc = float('inf')

            for cut in range(l, r, k-1):
                tc = compute(l, cut) + compute(cut+1, r)
                mc = min(mc, tc)
            

            if (r - l) % (k-1) == 0:
                mc += ic(l, r)
            
            return mc
        
        return compute(0, n-1)
```
