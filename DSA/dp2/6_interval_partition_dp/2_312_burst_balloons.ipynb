#### 312. Burst Balloons

* https://leetcode.com/problems/burst-balloons/

```python
from functools import lru_cache
from typing import List

# Invert the problem - Other than which ballon to burst first change it to which ballon to burst last
# if we know which ballon to burst last then we have two sublist of ballons on either side
# which are suppose to be burst first

# Same logic as lc 1547

class Solution:
    """
        Complexity	Value
        Time	O(n^3)
        Space	O(n^2)

        Why O(n^3)?

        O(n^2) intervals
        O(n) choices for last balloon
    """
    def maxCoins(self, nums: List[int]) -> int:
        """
            Returns maximum coins obtainable
            from balloons strictly inside
            interval (left, right).
        """
        values = [1] + [num for num in nums if num > 0] + [1]

        @lru_cache(maxsize=None)
        def solve(l, r):
            # No balloon exists inside interval.
            # Pruning
            if r - l <= 1:
                return 0

            max_coins = 0
            
            # Try each balloon as the LAST burst.
            for last_ballon in range(l+1, r):
                current_coins = values[last_ballon] * values[l] * values[r]

                total_coins = current_coins + solve(l, last_ballon) + solve(last_ballon, r)

                max_coins = max(max_coins, total_coins)
            
            return max_coins


        return solve(0, len(values)-1)
```

```python
from functools import lru_cache

class Solution:
    def maxCoins(self, nums) -> int:
        nums = [1] + nums + [1] # Add virtual balloons with value 1 at both ends
        n = len(nums)

        @lru_cache(None)
        def dfs(l ,r):
            if l+1 == r: # No balloon between left and right
                return 0

            max_coins = 0
            for i in range(l+1, r): # one at the left is avoided
                coins = nums[l] * nums[i] * nums[r] # burst i last
                coins += dfs(l, i) + dfs(i, r) # burst remaining
                max_coins = max(max_coins, coins)
            
            return max_coins
 
        return dfs(0, n-1) # 1 at the right is avoided

Solution().maxCoins([3,1,5,8])
```

🧠 Explanation:
Add 1 at the start and end of the array to handle edge balloons gracefully.

Use dp(left, right) to represent the maximum coins obtainable from bursting balloons between indices left and right exclusive.

For each possible last balloon to burst in the range (left, right), compute:

Coins earned from bursting it: nums[left] * nums[i] * nums[right]

Plus coins from solving the subproblems: dp(left, i) and dp(i, right)

Memoize results with lru_cache to avoid recomputation (top-down DP).
