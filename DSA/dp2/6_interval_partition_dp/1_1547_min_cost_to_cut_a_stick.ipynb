#### 1547. Minimum Cost to Cut a Stick

* https://leetcode.com/problems/minimum-cost-to-cut-a-stick/description/

```python
from typing import List
from functools import lru_cache

class Solution:
    """
    Top-Down Interval DP / Partition DP

    Idea:
    -----
    We recursively solve smaller stick segments and memoize results.

    For every interval:
        (positions[left], positions[right])

    we try every possible cut inside this interval.

    Cost of making a cut:
        length of current segment

    Total cost:
        current cut cost
        + left subproblem
        + right subproblem

    Recurrence:
    -----------
    dp(left, right) =
        segment_length
        + min(
            dp(left, cut_index)
            + dp(cut_index, right)
          )

    Time Complexity:
    ----------------
    O(m^3)

    Space Complexity:
    -----------------
    O(m^2)

    where:
        m = number of cuts + 2
    """
    def minCost(self, n: int, cuts: List[int]) -> int:
        positions = [0] + sorted(cuts) + [n]

        @lru_cache(maxsize=None)
        def solve(l, r):
            # this is used to get the csc from positions
            # and if r == l or r == l + 1
            # there won't be loop with ci
            # and mc will be infinity
            # hence this condition
            # no more cut possible as these are the boundaries defined in positions or cuts, using these we need to cut between these boundaries
            if r - l <= 1: 
                return 0
            curr_cost = positions[r] - positions[l]
            min_cost = float('inf')

            # why l+1, again what's the point cutting from l to l say 0 to 0 as it is boundary and nothing is in between
            # and the main reason is cut_index will be equal to l, so again we are solving solve(cut_index, r) i.e solve(l, r) 
            # so the problem never shrinks.
            for cut_index in range(l+1, r): 
                total_cost =  curr_cost + solve(l, cut_index) + solve(cut_index, r)
                min_cost = min(min_cost, total_cost)
            
            return min_cost

        return solve(0, len(positions)-1)
        
```
