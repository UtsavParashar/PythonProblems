##### 1039. Minimum Score Triangulation of Polygon

* https://leetcode.com/problems/minimum-score-triangulation-of-polygon/description/

```python
class Solution:
    def minScoreTriangulation(self, values: List[int]) -> int:
        """
            Pattern - Interval / Partition DP

            Any closed component with straight lines(and not curves) in a polygon
            Minimalist form of polygon is triangle 3 side - closed

            square can contain 2 triangles

            Logic - Cut the polygon into triangle from all possible points then return the min score
            Same as LC 1547, 312, 1039
            Follow standard template solution
        """
        @lru_cache(maxsize=None)
        def compute(l, r):
            if r - l < 2: # Triangle needs atleast 3 edges
                return 0
            
            min_score = float('inf')
            for cut in range(l+1, r):
                current_score = values[cut] * values[l] * values[r]
                total_score = current_score + compute(l, cut) + compute(cut, r)
                min_score = min(min_score, total_score)
            
            return min_score
        
        return compute(0, len(values)-1)

        
```
