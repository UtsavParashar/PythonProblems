##### 241. Different Ways to Add Parentheses

* https://leetcode.com/problems/different-ways-to-add-parentheses/description/

```python
class Solution:
    """
            There are O(n²) unique intervals due to memoization, but each interval can generate exponentially many expression outcomes. Since the algorithm enumerates all valid parenthesizations, the true complexity is governed by the Catalan number of binary expression trees:

            Time = O(Cn * n) ~ O(exponential)
            Space = O(Cn * n) ~ O(exponential)
    """
    def diffWaysToCompute(self, expression: str) -> List[int]:
        """
            Interval / Partition DP - cut at the operator

            Logic - Cut from the operator
            find the left results
            find the right results
            apply operator or left results and right results
        """ 

        @lru_cache(maxsize=None)
        def compute(l, r):
            res = []

            for i in range(l, r):
                operator = expression[i]
                if operator in {'+', '-', '*'}:
                    left_res = compute(l, i-1)
                    right_res = compute(i+1, r)

                    for left_val in left_res:
                        for right_val in right_res:
                            if operator == '+':
                                res.append(left_val + right_val)
                            elif operator == '-':
                                res.append(left_val - right_val)
                            elif operator == '*':
                                res.append(left_val * right_val)
            
            # Base case
            # if its not an operator but numbers
            if not res:
                res.append(int(expression[l: r+1]))
        
            return res
        
        return compute(0, len(expression)-1)
        
```
