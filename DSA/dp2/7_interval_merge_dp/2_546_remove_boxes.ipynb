#### 546. Remove Boxes

* https://leetcode.com/problems/remove-boxes/description/

```python
class Solution:
    """
        Time Complexity: $O(N^4)$ in the worst case. 
        The state space (l, r, k) is bounded by N * N * N, making it O(n^3) unique states. Inside the DP function, the for m loop iterates up to n times.
        
        Space Complexity: o(n^3) to store the memoization table (cache) for the states, 
        plus o(n) for the recursion stack depth.
    """
    def removeBoxes(self, boxes: List[int]) -> int:
        n = len(boxes)

        @cache
        def dfs(l, r, k):
            if l > r:
                return 0

            while l+1 <= r and boxes[l] == boxes[l+1]:
                l += 1
                k += 1
            
            # Branch 1 - Greedy - Cash out along the path
            # k is for the streak
            res = (k+1)**2 + dfs(l+1, r, 0)

            # Branch 2 - pin l, check for same element by skipping other elements
            # why l + 1 - because l and m element will be same so no point checking
            # why r + 1 - because we want to check the last element as well
            for m in range(l+1, r+1):
                # Imp condition - if same element is found after other elements
                if boxes[l] == boxes[m]:
                    # why - dfs(l+1, m-1) - because element at l and m are equal
                    # dfs(m, r) - because element at m and r could be equal
                    # k+ 1 - because streak increases
                    res = max(res, dfs(l+1, m-1, 0) + dfs(m, r, k + 1))
            
            return res
        
        return dfs(0, n-1, 0)
```

```python
class Solution:
    """
        Time Complexity: $O(N^4)$ in the worst case. 
        The state space (l, r, k) is bounded by N * N * N, making it O(n^3) unique states. Inside the DP function, the for m loop iterates up to n times.
        
        Space Complexity: o(n^3) to store the memoization table (cache) for the states, 
        plus o(n) for the recursion stack depth.
    """
    def removeBoxes(self, boxes: List[int]) -> int:
        n = len(boxes)

        @cache
        def dfs(l, r, k):
            if l > r:
                return 0

            ## Optimization: Fast-forward through contiguous identical boxes
            while l+1 <= r and boxes[l] == boxes[l+1]:
                l += 1
                k += 1

            # Strategy 1 - Cash out the current contiguous block
            # eg 8 9 9 8 = 1^2 + 2^2 + 1^2 = 6
            max_profit = (k+1) * (k+1) + dfs(l+1, r, 0)

            # Strategy 2: Defer and merge with a future identical box
            for m in range(l+1, r+1):
                if boxes[l] == boxes[m]:
                    # Clear the middle boxes (dp(l + 1, m - 1, 0))
                    # Merge current block with the future matching box (dp(m, r, k + 1))
                    intermediate_profit = dfs(l+1, m-1, 0)
                    combined_profit = dfs(m, r, k+1)

                    max_profit = max(max_profit, intermediate_profit + combined_profit)
            
            return max_profit
        
        return dfs(0, n-1, 0)
```
