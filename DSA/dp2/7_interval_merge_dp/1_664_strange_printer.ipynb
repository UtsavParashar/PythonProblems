```python
class Solution:
    def strangePrinter(self, s: str) -> int:
        compressed_text = []

        # Optimization: Consecutive duplicate characters are redundant
        # Eg. "aaabbbcc" -> "abc"
        # The printer can already print continuous identical characters in a single turn.
        for c in s:
            if not compressed_text or compressed_text[-1] != c:
                compressed_text.append(c)

        text = ''.join(compressed_text)

        n = len(text)

        @cache
        def dfs(l, r):
            if l > r:
                return 0
            
            if l == r:
                return 1

            mt = 1+ dfs(l+1, r)

            for m in range(l+1, r+1):
                if text[l] == text[m]:
                    middle_turns = dfs(l+1, m-1)
                    remaining_turns = dfs(m, r)

                    mt = min(mt, middle_turns + remaining_turns)
            
            return mt
        
        return dfs(0, n-1)
```
