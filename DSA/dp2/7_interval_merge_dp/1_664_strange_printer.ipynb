#### 664. Strange Printer
* https://leetcode.com/problems/strange-printer/description/

```python
class Solution:
    def strangePrinter(self, s: str) -> int:
        
        # why compressed_text
        # if two consecutive elements are already same 
        # then they don't need to be counted for printing twice
        compressed_text = []

        for c in s:
            if not compressed_text or compressed_text[-1] != c:
                compressed_text.append(c)

        s = ''.join(compressed_text)
        n = len(s)

        @cache
        def dfs(l, r):
            if l > r:
                return 0
            if l == r:
                return 1

            # Branch 1 - Greedily cash out at every step
            res = 1 + dfs(l+1, r)

            # Branch 2 - Form the stream to see if more than one element can printed in one go
            for m in range(l+1, r+1):
                if s[l] == s[m]:
                    res = min(res, dfs(l+1, m-1) + dfs(m, r))
            
            return res
        
        return dfs(0, n-1)
            
```
