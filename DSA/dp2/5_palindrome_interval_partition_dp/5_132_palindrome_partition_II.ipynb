#### 132. Palindrome Partitioning II
* https://leetcode.com/problems/palindrome-partitioning-ii/description/

```python
# Preferred SOlution

class Solution:
    def minCut(self, s: str) -> int:
        """
            Preferred solution - 
            TC - O(n^2)
            SC - O(n)

            1D DP for storing the min number of cuts

            why min_cut_dp[r] = 0 if l == 0: its the base case which defines that no cuts are required eg aabaa - 0 cuts are req
            
            why min_cut_dp[r] = min(min_cut_dp[r], min_cut_dp[l-1] + 1) :
            previous optimal solution till l is in min_cut_dp[l-1], +1 tells one cut is required after that
            min_cut_dp[r] - Cuts required till r
            since we need min of both hence min function over it

            Eg aa bcb -> till aa no cut hence optimal boundary till l-1 is 0 cuts
            then bcb is a palindrome so no cut is required

            why min_cut_dp[r] and not min_cut_dp[l] -> at each step we are finding min cuts required till r
            i.e right boundary of prefix as the problem could be stated as - 
            "What is the minimum cut needed for prefix ending at RIGHT?"
           Each state has ONE precise meaning. ->  minimum_cuts[i]=minimum cuts for s[0:i]

           Real world analogy in electronic system
           Big sell order of 1m shares, 
           strategy applied Symmetric mean reversion of liquidity ie if liqudity is a high sell at a venue else dark pool
           HHLLHH -> can be treated as one self-contained liquidity regime hence 0 cuts
           Meaning:
                do NOT switch execution strategy mid-way
                preserve execution continuity
                avoid excessive venue reconfiguration
        """
        n = len(s)
        min_cut_dp = [i for i in range(n)]

        def expand_pal(l, r):
            while l >= 0 and r < n and s[l] == s[r]:
                if l == 0:
                    min_cut_dp[r] = 0
                else:
                    min_cut_dp[r] = min(min_cut_dp[r], min_cut_dp[l-1] + 1)
                l -= 1
                r += 1
            
        for i in range(n):
            expand_pal(i, i)
            expand_pal(i, i+1)
        
        return min_cut_dp[-1]
```

```python
from functools import lru_cache
class Solution:
    """
    Top-Down Memoized DFS Solution

    Time Complexity:
        O(n^2)

    Space Complexity:
        O(n^2)

    Design Notes:
        - Memoized recursion
        - Precomputed palindrome table
        - Clean state decomposition
        - Interview friendly
        - Production readable
    """
    def minCut(self, s: str) -> int:
        """
            Type - Palindrome DP, Interval Partition DP
            Can be divided into two parts -
            1. Find if the substring in palindrome
            Solve it using 2D matrix because this will make loopup as O(1)
            else TLE will be raised
            2. check for all substrings
        """
        n = len(s)

        # ---------------------------------------------------------
        # Step 1:
        # Precompute palindrome substrings
        # ---------------------------------------------------------
        dp = [[False]*n for _ in range(n)]

        for l in range(n-1, -1, -1):
            for r in range(l, n):
                if s[l] == s[r]:
                    if r-l <= 2 or dp[l+1][r-1]: # Revisit this logic to understand
                        dp[l][r] = True
        
        # ---------------------------------------------------------
        # Step 2:
        # DFS + Memoization
        # ---------------------------------------------------------
        @lru_cache(maxsize=None)
        def mp(si):
            if si == n:
                return 0
            
            mcp = float('inf')

            for ei in range(si, n):
                if dp[si][ei]:
                    cpc = mp(ei+1) + 1
                    mcp = min(mcp, cpc)
            return mcp
    
        # partitions - 1 = cuts
        return mp(0) - 1
```

```python
def minCut(s: str) -> int:
    n = len(s)
    # Precompute palindrome table
    is_pal = [[False] * n for _ in range(n)]
    for end in range(n):
        for start in range(end + 1):
            if s[start] == s[end] and (end - start <= 2 or is_pal[start + 1][end - 1]):
                is_pal[start][end] = True

    # DP for min cuts
    dp = [0] * n
    for i in range(n):
        if is_pal[0][i]:
            dp[i] = 0
        else:
            dp[i] = min(dp[j - 1] + 1 for j in range(1, i + 1) if is_pal[j][i])
    return dp[-1]

```

```python
# TLE
# Time: O(n³) in the worst case without palindrome precomputation (since is_pal check is O(n) each call).
# Space: O(n) recursion depth + O(n) memo table.

from functools import lru_cache

class Solution:
    def minCut(self, s: str) -> int:
        n = len(s)

        def is_pal(l, r):
            while l < r:
                if s[l] != s[r]:
                    return False
                l += 1
                r -= 1
            return True

        @lru_cache(None)
        def dfs(i):
            # If reached end or remaining substring is palindrome → no cuts needed
            if i == n or is_pal(i, n-1):
                return 0

            min_cuts = float('inf')
            for j in range(i, n):
                if is_pal(i, j):
                    # 1 cut + solve remainder
                    min_cuts = min(min_cuts, 1+dfs(j+1))

            return min_cuts

        return dfs(0)
Solution().minCut('aab')
```
