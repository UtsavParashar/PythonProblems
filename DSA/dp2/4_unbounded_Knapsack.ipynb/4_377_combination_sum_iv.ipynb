#### 377. Combination Sum IV

* https://leetcode.com/problems/combination-sum-iv/description/

```python
class Solution:
    """
    Unbounded Knapsack
    This is a permutations problem disguised as Combination
    Permutation has more resultant values
    because 1,2 != 2,1, both are different

    Since this is a permutation problem 
    and we could reuse smaller num after greater num
    so better to use target as the outer for loop and 
    num as the inner for loop

    This problem is similar to lc 518
    """
    def combinationSum4(self, nums, target: int) -> int:
        """
        Dynamic Programming:
        dp[total] = number of ordered sequences that form 'total'

        Time Complexity:
            O(target * len(nums))

        Space Complexity:
            O(target)
        """
        dp = [0]*(target+1)
        # One way to make 0 -> choose nothing
        dp[0] = 1
        
        for current_target in range(1, target+1):
            for num in nums:
                if current_target >= num:
                    dp[current_target] += dp[current_target-num]
        
        return dp[target]
```
