#### 322. Coin Change

* https://leetcode.com/problems/coin-change/description/

```python
class Solution:
    """
        # Here coins is good to have at the outer loop
        # because its a combination problem where
        # sequence does not matter that is 1,2 == 2, 1 == 3$
        # however if it is a knapsack unbounded permutations problem like 
        # 377. Combination Sum IV then 
        # amount/target needs to be the outer loop becasue 1,2 != 2,1 in that case. 
    """
    def coinChange(self, coins, amount: int) -> int:
        if len(coins) == 0 or amount == 0:
            return 0
        
        INF = amount + 1

        dp = [INF]*(amount + 1)
        dp[0] = 0

        for coin in coins:
            for current_sum in range(coin, amount + 1):
                dp[current_sum] = min(dp[current_sum], 1+dp[current_sum - coin])

        return dp[amount] if dp[amount] != INF else -1
```
