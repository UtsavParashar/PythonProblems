##### 416. Partition Equal Subset Sum

```python
from typing import List

class Solution:
    """
        Determines if the input array can be partitioned into two subsets
        with equal sum.

        Approach:
        - Convert to subset sum problem (target = total_sum // 2)
        - Use space-optimized 1D DP (boolean array)
    """
    def canPartition(self, nums: List[int]) -> bool:
        """
            Subsequence DP (Pick/Not Pick)
            TC - O(n * target)
            SC - O(target)

            Logic - 
            1. sum can be divided in two ints only if it is event
            2. since we need to divided by 2 so target = total sum by 2
            3. dp state defined if we can get target by getting sub targets sum
            4. inner loop moves in opposite dir to avoid duplication
        """

        total_sum = sum(nums)
        target = total_sum // 2

        if total_sum % 2 != 0:
            return False

        dp = [False] * (target+1)
        dp[0] = True # Base case as 0 can be divided into two equal halves as 0

        for num in nums:
            # Traverse backwards to avoid reuse of same element
            for current_sum in range(target, num-1, -1):
                if dp[current_sum - num]:
                    dp[current_sum] = True
            
            # Early stopping important optimization in interviews
            if dp[target]:
                return True
        
        return dp[target]
```
