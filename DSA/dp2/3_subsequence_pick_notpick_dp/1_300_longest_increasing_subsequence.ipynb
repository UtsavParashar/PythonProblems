##### 300. Longest Increasing Subsequence

* https://leetcode.com/problems/longest-increasing-subsequence/description/

```python
from typing import List
import bisect

class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        """
            Preferred Solution - Using Greedy + Binary Search
             O(n log n), O(n)
            Alternate Solution - Subsequence(Pick/Not Pick) DP
        """
        if not nums:
            return 0

        tails = []

        for num in nums:
            idx = bisect.bisect_left(tails, num)
            if idx == len(tails):
                tails.append(num)
            else:
                tails[idx] = num
        
        return len(tails)
```

```python
from typing import List

class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        """
            Alternate Solution - Subsequence(Pick/Not Pick) DP
            O(n^2), O(n)

            At each index i, check all previous indexes from inception that element at i is greater than how many of them
        """
        n = len(nums)
        dp = [1]*n

        for i in range(n):
            for j in range(i):
                if nums[i] > nums[j]:
                    dp[i] = max(dp[i], dp[j]+1)
        
        return max(dp, default=0)
```
