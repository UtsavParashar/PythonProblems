#### 647. Palindromic Substrings

* https://leetcode.com/problems/palindromic-substrings/description/

```python
# Solve palindrome problems in this order
# 1. lc 5. Longest Palindromic Substring - 2 pointers
# 2. lc 647. Palindromix Substrings - 2 pointers
# 3. lc 516. Londest Palindromic Subsequence - Palindrome (Interval) DP / Subsequence DP (lc 1143. Longest Common Subsequence)

class Solution:
    def countSubstrings(self, s: str) -> int:
        """
        LeetCode 647 - Palindromic Substrings

        Strategy:
        Expand around every possible palindrome center.

        Time Complexity:
            O(n^2)

        Space Complexity:
            O(1)
        """
        
        n = len(s)

        def is_pal(l, r):
            discovered_palindromes  = 0
            while l >= 0 and r < n and s[l] == s[r]:
                discovered_palindromes  += 1
                l -= 1
                r += 1
            return discovered_palindromes 
        
        res = 0
        for i in range(n):
            res += is_pal(i, i)
            res += is_pal(i, i+1)
        
        return res
```
