#### 1209. Remove All Adjacent Duplicates in String II

* https://leetcode.com/problems/remove-all-adjacent-duplicates-in-string-ii/description/

#### BBG - IMP

```python
class Solution:
    def removeDuplicates(self, s: str, k: int) -> str:
        """
        Removes all k-adjacent duplicates from the string.

        Time Complexity: O(n)
        Space Complexity: O(n)

        This implementation is optimized for clarity, correctness,
        and interview-readiness (MAANG / Bloomberg style).
        """

        # Stack holds pairs of (character, consecutive_count)
        stack = []

        for char in s:
            # Case 1: Same as previous character → increment count
            if stack and stack[-1][0] == char:
                stack[-1][1] += 1

                # Remove when count reaches k
                if stack[-1][1] == k:
                    stack.pop()
            else:
                # Case 2: New character → push with count 1
                stack.append([char, 1])

        # Rebuild the string from stack
        result = []
        for char, count in stack:
            result.append(char * count)

        return "".join(result)
```
