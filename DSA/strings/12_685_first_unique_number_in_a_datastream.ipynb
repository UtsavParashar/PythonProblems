```python
#### Confirm with interviewer if they say to imitate data stream 
#### and not consider the nums in memory then use queue based solution


from typing import List
from collections import Counter

class Solution:
    """
    TC - O(n)
    SC - O(n)
    """
    def first_unique_number(self, nums: List[int], number: int) -> int:
        # Write your code here
        freq_counter = Counter()
        end_index = 0

        for num in nums:
            freq_counter[num] += 1
            end_index += 1

            if num == number:
                break
        else:
            return -1

        for i in range(end_index+1):
            if freq_counter[nums[i]] == 1:
                return nums[i]

        return -1 

Solution().first_unique_number([1, 2, 2, 1, 3, 4], 3)
```

```python
#### QUEUE SOLUTION
from typing import List
from collections import Counter, deque

class Solution:
    """
    TC - O(n)
    SC - O(n)
    """
    def first_unique_number(self, nums: List[int], number: int) -> int:
        # Write your code here
        freq_counter = Counter()
        queue = deque()

        for num in nums:
            freq_counter[num] += 1
            queue.append(num)

            if num == number:
                break
        else:
            return -1

        while queue:
            if freq_counter[queue[0]] == 1:
                return queue[0]
            queue.popleft()

        return -1 

Solution().first_unique_number([1, 2, 2, 1, 3, 4], 3)
```
