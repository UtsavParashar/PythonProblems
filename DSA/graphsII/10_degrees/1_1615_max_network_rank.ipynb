##### 1615. Maximal Network Rank

* https://leetcode.com/problems/maximal-network-rank/description/

```python
from typing import List, Set, Tuple
from collections import defaultdict


class Solution:
    """
    Optimized solution using degree-based pruning.

    Key Idea:
    ----------
    Instead of checking all O(n^2) pairs, we focus only on nodes
    with highest and second-highest degrees.

    Time Complexity: O(n + m)
    Space Complexity: O(n + m)
    """

    def maximalNetworkRank(self, n: int, roads: List[List[int]]) -> int:
        # Step 1: Degree array
        degree = [0] * n

        # Step 2: Adjacency set for O(1) edge check
        adjacency: List[Set[int]] = [set() for _ in range(n)]

        for u, v in roads:
            degree[u] += 1
            degree[v] += 1
            adjacency[u].add(v)
            adjacency[v].add(u)

        # Step 3: Group nodes by degree
        degree_to_nodes = defaultdict(list)
        for node in range(n):
            degree_to_nodes[degree[node]].append(node)

        # Step 4: Sort degrees descending
        sorted_degrees = sorted(degree_to_nodes.keys(), reverse=True)

        max_degree = sorted_degrees[0]
        max_nodes = degree_to_nodes[max_degree]

        # Case 1: Multiple nodes with max degree
        if len(max_nodes) > 1:
            # Check if any pair is NOT connected
            for i in range(len(max_nodes)):
                for j in range(i + 1, len(max_nodes)):
                    u, v = max_nodes[i], max_nodes[j]
                    if v not in adjacency[u]:
                        return 2 * max_degree

            # All pairs are connected
            return 2 * max_degree - 1

        # Case 2: Only one node with max degree
        u = max_nodes[0]
        second_max_degree = sorted_degrees[1]
        second_nodes = degree_to_nodes[second_max_degree]

        for v in second_nodes:
            if v not in adjacency[u]:
                return max_degree + second_max_degree

        return max_degree + second_max_degree - 1
```

```python
from typing import List, Set, Tuple


class Solution:
    """
    LeetCode 1615: Maximal Network Rank

    Approach:
    ----------
    1. Compute degree (number of connections) for each city.
    2. Store all direct connections in a hash set for O(1) lookup.
    3. Iterate over all pairs of cities (u, v) where u < v:
        - Network rank = degree[u] + degree[v]
        - Subtract 1 if u and v are directly connected
    4. Track the maximum network rank.

    Time Complexity: O(n^2 + m)
    Space Complexity: O(n + m)
    """

    def maximalNetworkRank(self, n: int, roads: List[List[int]]) -> int:
        # Step 1: Degree array
        degree: List[int] = [0] * n

        # Step 2: Edge set for quick lookup
        edge_set: Set[Tuple[int, int]] = set()

        # Build degree array and edge set
        for u, v in roads:
            degree[u] += 1
            degree[v] += 1

            # Store edges in normalized order (min, max)
            edge_set.add((min(u, v), max(u, v)))

        max_rank: int = 0

        # Step 3: Check all pairs
        for city_a in range(n):
            for city_b in range(city_a + 1, n):
                # Base rank = sum of degrees
                current_rank: int = degree[city_a] + degree[city_b]

                # If directly connected, subtract 1
                if (city_a, city_b) in edge_set:
                    current_rank -= 1

                # Update maximum
                if current_rank > max_rank:
                    max_rank = current_rank

        return max_rank
```
