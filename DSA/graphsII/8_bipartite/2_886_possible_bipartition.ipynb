#### 886. Possible Bipartition

* https://leetcode.com/problems/possible-bipartition/description/

```python
from collections import deque
from typing import List


class Solution:
    def possibleBipartition(self, num_people: int, dislikes: List[List[int]]) -> bool:
        """
        Inline BFS version without helper function.

        Time Complexity: O(N + E)
        Space Complexity: O(N + E)
        """

        # Build adjacency list
        adjacency_list = [[] for _ in range(num_people + 1)]
        for person_a, person_b in dislikes:
            adjacency_list[person_a].append(person_b)
            adjacency_list[person_b].append(person_a)

        # 0 = unvisited, 1 / -1 = groups
        color = [0] * (num_people + 1)

        # Process all components
        for person in range(1, num_people + 1):
            if color[person] != 0:
                continue

            # Start BFS inline
            queue = deque([person])
            color[person] = 1

            while queue:
                current_person = queue.popleft()

                for neighbor in adjacency_list[current_person]:
                    if color[neighbor] == 0:
                        color[neighbor] = -color[current_person]
                        queue.append(neighbor)
                    elif color[neighbor] == color[current_person]:
                        return False

        return True
```
