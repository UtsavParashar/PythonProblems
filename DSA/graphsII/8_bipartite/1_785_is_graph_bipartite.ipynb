# 785. Is Graph Bipartite?

* https://leetcode.com/problems/is-graph-bipartite/description/?/

# IMP

```python
from typing import List
from collections import deque
class Solution:
    """
        Determines if an undirected graph is bipartite.

        Args:
            graph: Adjacency list representation where graph[u] contains neighbors of node u.

        Returns:
            True if graph is bipartite, False otherwise.

        Approach:
            - Use BFS to attempt 2-coloring of the graph.
            - Handle disconnected components by iterating over all nodes.

    * Each node visited once
    * Each edge processed twice (undirected)
    """
    def is_bipartite(self, graph: List[List[int]]) -> bool:
        """
            Bipartite Graph is a graph which could be segregated into two groups.
            Nodes in one group connects to the nodes in other group.
            There could be one to one and one to many and many to one relationship.
            We could color the nodes as 0 and 1 to segregate the groups.

            A graph is not Bipartite.
            If a node points to itself.
            If two nodes in a same group are connected forming odd cycle.

            This can be solved with both BFS/DFS. BFS preferred.

            TC - O(V+E)
            SC - O(V)
        """

        n = len(graph)

        # -1 = uncolored, 0 and 1 are the two colors
        colors = [-1]*n

        # this loop is required if multiple graphs are not connected
        for start_node in range(n):
            # Skip already visited components
            if colors[start_node] != -1:
                continue
            
            # Start BFS for this component
            q = deque([start_node])
            colors[start_node] = 0
            while q:
                curr_node = q.popleft()

                for nei in graph[curr_node]:
                    if colors[nei] == -1:
                        colors[nei] = 1 - colors[curr_node]
                        q.append(nei)
                    
                    # if node and its neighbour has same value that means they are adjacent
                    elif colors[nei] == colors[curr_node]: 
                        return False
        
        return True


Solution().is_bipartite([[1,2,3],[0,2],[0,1,3],[0,2]]) # negative case
Solution().is_bipartite([[1,3],[0,2],[1,3],[0,2]]) # positive case
Solution().is_bipartite([[1],[0],[3],[2]]) # two unconnected graphs
Solution().is_bipartite([[2, 3],[],[0],[0]]) # two unconnected graphs
```
