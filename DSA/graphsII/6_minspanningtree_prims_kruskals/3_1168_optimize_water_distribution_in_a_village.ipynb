#### 1168. Optimize Water Distribution in a Village

* https://algo.monster/liteproblems/1168

```python
from typing import List


class UnionFind:
    """
    Disjoint Set Union (Union-Find) with:
    - Path Compression
    - Union by Rank
    - Component tracking
    """
    def __init__(self, size: int) -> None:
        self.parent = list(range(size))
        self.rank = [0] * size
        self.components = size  # useful for early stopping

    def find(self, node: int) -> int:
        """Find root with path compression."""
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        return self.parent[node]

    def union(self, u: int, v: int) -> bool:
        """
        Union by rank.
        Returns True if union happened (i.e., different components).
        """
        root_u = self.find(u)
        root_v = self.find(v)

        if root_u == root_v:
            return False

        # Attach smaller rank tree under larger
        if self.rank[root_u] < self.rank[root_v]:
            self.parent[root_u] = root_v
        elif self.rank[root_u] > self.rank[root_v]:
            self.parent[root_v] = root_u
        else:
            self.parent[root_v] = root_u
            self.rank[root_u] += 1

        self.components -= 1
        return True


class Solution:
    def minCostToSupplyWater(
        self,
        n: int,
        wells: List[int],
        pipes: List[List[int]]
    ) -> int:
        """
        Kruskal's MST with virtual node (0).
        
        Time Complexity: O(E log E)
        Space Complexity: O(N + E)
        """

        # Step 1: Build edges list
        edges = []

        # Add virtual edges: (cost, 0, house)
        for house in range(1, n + 1):
            edges.append((wells[house - 1], 0, house))

        # Add pipe edges
        for house1, house2, cost in pipes:
            edges.append((cost, house1, house2))

        # Step 2: Sort edges by cost
        edges.sort(key=lambda edge: edge[0])

        # Step 3: Initialize Union-Find
        uf = UnionFind(n + 1)  # include virtual node 0

        total_cost = 0

        # Step 4: Kruskal's Algorithm
        for cost, u, v in edges:
            if uf.union(u, v):
                total_cost += cost

                # Early stopping: all nodes connected
                if uf.components == 1:
                    break

        return total_cost
```
