#### 1135. Connecting Cities With Minimum Cost

* https://algo.monster/liteproblems/1135

```python
# https://chatgpt.com/c/69c0e0a0-a748-8321-ae83-374153addbe7
### Use Kruskal's algo when graph is sparse and Prims when graph is dense
### Kruskal time complexity - E log E
### if graph is dense, E ~ V^2 so it becomes E log V^2, each node connected to almost every other node
### Prims time complexity - E log V

### As a rule of thumb, if edge weights are given use Kruskal's, unless specified that the graph is dense use Kruskals else Prims
```

```python
#### Kruskals algo - Sort the edges  by cost + Union Find
### You can either used edges_used or self.components, however in MST edges_used in preferred
class UnionFind:
    def __init__(self, size):
        self.parent = list(range(size+1))
        self.rank = [0]*(size+1)
    
    # path compression
    def find(self, node):
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        return self.parent[node]
    
    # union by rank
    def union(self, nodea, nodeb):
        roota = self.find(nodea)
        rootb = self.find(nodeb)

        if roota == rootb:
            return False

        if self.rank[roota] < self.rank[rootb]:
            self.parent[roota] = rootb
        elif self.rank[roota] > self.rank[rootb]:
            self.parent[rootb] = roota
        else:
            self.parent[rootb] = roota
            self.rank[roota] += 1
        

        return True



class Solution:
    def min_cost(self, n, connections):
        """
        Returns the minimum cost to connect all cities using MST.
        If not possible, returns -1.
        """

        connections.sort(key=lambda edge: edge[2])

        uf = UnionFind(n)

        total_cost = 0
        edges_used = 0
        for city1, city2, cost in connections:
            if uf.union(city1, city2):
                total_cost += cost
                edges_used += 1
            
            if edges_used == n-1:
                return total_cost
        
        return total_cost if edges_used == n-1 else -1

Solution().min_cost(3, [[1,2,5], [1,3,6], [2,3,1]])
```
