#### 207. Course Schedule

* https://leetcode.com/problems/course-schedule/description/

### IMP

```python
from collections import deque, defaultdict
from typing import List


class Solution:
    """
    Determine whether all courses can be completed given prerequisite dependencies.

    This solution uses Kahn's Algorithm (BFS-based Topological Sort) to detect
    cycles in a directed graph.

    If a cycle exists, a valid topological ordering is impossible,
    meaning some courses cannot be completed.
    """

    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        """
        Args:
            numCourses: Total number of courses (nodes in graph)
            prerequisites: List of prerequisite pairs [course, prerequisite]

        Returns:
            True if all courses can be completed (no cycle)
            False otherwise

        Time Complexity:  O(V + E)
        Space Complexity: O(V + E)
        """

        # -----------------------------
        # Step 1: Build graph structure
        # -----------------------------
        dag = defaultdict(list)
        indegree = [0]*numCourses
        for dep, indep in prerequisites:
            dag[indep].append(dep)
            indegree[dep] += 1

        # --------------------------------------
        # Step 2: Initialize queue with sources
        # (courses having no prerequisites)
        # --------------------------------------
        queue = deque(course for course in range(numCourses) if indegree[course] == 0)

        completed_courses = 0

        # --------------------------------------
        # Step 3: BFS Topological traversal
        # --------------------------------------
        while queue:
            curr_course = queue.popleft()
            completed_courses += 1

            for dep_course in dag[curr_course]:
                indegree[dep_course] -= 1
                if indegree[dep_course] == 0:
                    queue.append(dep_course)

        # --------------------------------------
        # Step 4: If all nodes processed → no cycle
        # --------------------------------------
        return completed_courses == numCourses
```
