#### 1136. Parallel Courses

* https://algo.monster/liteproblems/1136

### IMP

```python
from collections import defaultdict, deque
class Solution:
    """
    Solves LeetCode 1136 - Parallel Courses

    Uses Kahn's Algorithm (Topological Sort via BFS) where each BFS level
    represents one semester of courses that can be taken in parallel.

    TC and SC - O(V+E)
    """
    def minimun_sequence(self, n, relations): # relations = [prevCourse, nextCourse]
        """
        Args:
            n: Total number of courses labeled from 1..n
            relations: prerequisite pairs [prev_course, next_course]

        Returns:
            Minimum semesters needed to finish all courses.
            Returns -1 if impossible due to cyclic dependency.
        """
        dag = defaultdict(list)
        indegree = [0]*(n+1)

        for prev_course, next_course in relations:
            dag[prev_course].append(next_course)
            indegree[next_course] += 1

        queue = deque()
        for course in range(1, n+1):
            if indegree[course] == 0:
                queue.append(course)

        semesters = 0
        courses_completed = 0
        while queue:
            semesters += 1
            for _ in range(len(queue)):
                curr_course = queue.popleft()
                courses_completed += 1

                for next_course in dag[curr_course]:
                    indegree[next_course] -= 1
                    if indegree[next_course] == 0:
                        queue.append(next_course)
            
        # cycle detected, circular dependency or all courses not completed
        return semesters if courses_completed == n else -1

Solution().minimun_sequence(5, relations = [[1,3], [2,3], [3,4], [3,5]])
```
