#### 802. Find Eventual Safe States

* https://leetcode.com/problems/find-eventual-safe-states/

```python
# Simple DAG problem
# which requires outdegree + rev graph
# we maintain res[F] list for the sorted result in output so that other than sorting the res at the end nlogn operation
# we are using o(n)space and o(n) time
# TC - O(V+E)
# SC - O(V+E)

from typing import List
from collections import defaultdict, deque

class Solution:
    """
    Computes all eventual safe nodes in a directed graph.

    Approach:
    - Reverse graph construction
    - Track outdegree of each node
    - BFS (Kahn’s Algorithm) from terminal nodes

    Time Complexity: O(V + E)
    Space Complexity: O(V + E)
    """
    def eventualSafeNodes(self, graph: List[List[int]]) -> List[int]:
        n = len(graph)
        rev_graph = defaultdict(list)
        outdegree = [0]*n

        for i in range(n):
            if graph[i]:
                for node in graph[i]:
                    rev_graph[node].append(i)
                    outdegree[i] += 1
        
        queue = deque(i for i in range(n) if outdegree[i] == 0)
        res = [False]*n
        while queue:
            cn = queue.popleft()
            res[cn] = True

            for nei in rev_graph[cn]:
                outdegree[nei] -= 1
                if outdegree[nei] == 0:
                    queue.append(nei)
        
        return [node for node in range(n) if res[node]]
```

```python
# Ref - https://www.youtube.com/watch?v=wciKkM3g3wQ

class Solution:
    def eventualSafeNodes(self, graph):
        safe_map = {}
        res = []

        n = len(graph)

        def dfs(node):
            if node in safe_map:
                return safe_map[node]

            safe_map[node] = False

            for nei in graph[node]:
                if not dfs(nei):
                    return False
            
            safe_map[node] = True
            return safe_map[node]

        for i in range(n):
            if dfs(i):
                res.append(i)

        return res

Solution().eventualSafeNodes(graph = [[1,2],[2,3],[5],[0],[5],[],[]])
```
