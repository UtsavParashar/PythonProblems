#### 684. Redundant Connection

* https://leetcode.com/problems/redundant-connection/description/

```python
class UnionFind:
    """
    Production-grade Union-Find (Disjoint Set Union).

    Optimizations:
    1. Path Compression
    2. Union by Rank

    Both optimizations ensure near constant-time operations.
    """
    def __init__(self, size):
        self.parent = list(range(size+1))
        self.rank = [0]*(size+1)

    def find(self, node):
        """
        Finds root of the set containing 'node'
        with path compression.
        """
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        return self.parent[node]

    def union(self, nodea, nodeb):
        """
        Unions the sets of node_a and node_b.

        Returns:
            True  -> union successful
            False -> nodes already connected (cycle detected)
        """
        roota = self.find(nodea)
        rootb = self.find(nodeb)

        if roota == rootb:
            return False

        if self.rank[roota] < self.rank[rootb]:
            self.parent[roota] = rootb
        elif self.rank[roota] > self.rank[rootb]:
            self.parent[rootb] = roota
        else:
            self.parent[rootb] = roota
            self.rank[roota] += 1
        
        return True

class Solution:
    """
    Solves Redundant Connection using Union-Find.

    Detects the edge that introduces a cycle
    in an otherwise tree-like graph.
    TC, SC = O(n)
    """
    def findRedundantConnection(self, edges: List[List[int]]) -> List[int]:
        n = len(edges)

        union_find = UnionFind(n)

        for src, dest in edges:
            if not union_find.union(src, dest):
                return [src, dest]
        
        return []
        
```
