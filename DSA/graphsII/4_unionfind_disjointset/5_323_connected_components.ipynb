#### 323. Number of Connected Components in an Undirected Graph

* https://neetcode.io/problems/count-connected-components/question

### IMP

```python
class UnionFind:
    """
    LeetCode 323 - Number of Connected Components in an Undirected Graph

    Uses Union-Find to merge connected nodes.

    Time Complexity:
        O(E * α(N))

    Space Complexity:
        O(N)

    where α(N) is the inverse Ackermann function
    which grows extremely slowly (≈ constant).
    """
    def __init__(self, size):
        self.parent = list(range(size))
        self.rank = [0]*size
        self.connected_components = size

    def find(self, node):
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        return self.parent[node]

    def union(self, nodea, nodeb):
        roota = self.find(nodea)
        rootb = self.find(nodeb)

        if roota == rootb:
            return False
        
        if self.rank[roota] < self.rank[rootb]:
            self.parent[roota] = rootb
        elif self.rank[roota] > self.rank[rootb]:
            self.parent[rootb] = roota
        else:
            self.parent[rootb] = roota
            self.rank[roota] += 1
        
        self.connected_components -= 1
        return True

class Solution:
    def countComponents(self, n: int, edges: List[List[int]]) -> int:

        uf = UnionFind(n)
        for u, v in edges:
            uf.union(u, v)
        
        return uf.connected_components
        
```
