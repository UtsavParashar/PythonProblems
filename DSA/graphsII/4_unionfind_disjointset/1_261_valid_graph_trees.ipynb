### Graph Valid Tree

* https://neetcode.io/problems/valid-tree/question

```python
class UnionFind:
    """
    Disjoint Set Union (Union-Find) with
    path compression and union by rank.

    Amortized complexity ~ O(n).
    """
    def __init__(self, size):
        self.parent = list(range(size))
        self.rank = [0]*size

    # Path compression - flatten the tree
    # all childs become direct child or root
    def find(self, node):
        """
        Finds the root parent of a node
        using path compression.
        """
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        return self.parent[node]

    def union(self, nodea, nodeb):
        """
        Union two sets.

        Returns:
            False if both nodes already share the same parent (cycle detected)
            True otherwise
        """
        roota = self.find(nodea)
        rootb = self.find(nodeb)

        # cycle detected
        if roota == rootb:
            return False

        # Union by rank
        # adding smaller tree as a child of bigger tree
        # this will be better than adding bigger tree as a child 
        # of smaller tree in terms of height
        if self.rank[roota] < self.rank[rootb]:
            self.parent[roota] = rootb
        elif self.rank[roota] > self.rank[rootb]:
            self.parent[rootb] = roota
        else:
            self.parent[rootb] = roota
            self.rank[roota] += 1

        return True


class Solution:
    """
        A graph is a valid tree if:
        1. It has exactly n-1 edges
        2. It contains no cycles
        TC, SC - O(n)
    """
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        # Optimization: Tree must have exactly n-1 edges
        if len(edges) != n-1:
            return False

        union_find = UnionFind(n)

        for src, dest in edges:
            if not union_find.union(src, dest): # Cycle detected
                return False

        return True
        
```
