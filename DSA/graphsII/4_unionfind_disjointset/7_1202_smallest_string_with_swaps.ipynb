#### 1202. Smallest String With Swaps

* https://leetcode.com/problems/smallest-string-with-swaps/description/

```python
from collections import defaultdict
from typing import List


class UnionFind:
    """
    Disjoint Set (Union-Find) with:
        - Path Compression
        - Union by Rank

    Amortized Time Complexity:
        find  -> O(α(n))
        union -> O(α(n))
    """

    def __init__(self, size: int) -> None:
        self.parent = list(range(size))
        self.rank = [0] * size

    def find(self, node: int) -> int:
        """
        Find root of node with path compression.
        """
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        return self.parent[node]

    def union(self, node_a: int, node_b: int) -> None:
        """
        Union two components by rank.
        """
        root_a = self.find(node_a)
        root_b = self.find(node_b)

        if root_a == root_b:
            return

        if self.rank[root_a] < self.rank[root_b]:
            self.parent[root_a] = root_b
        elif self.rank[root_a] > self.rank[root_b]:
            self.parent[root_b] = root_a
        else:
            self.parent[root_b] = root_a
            self.rank[root_a] += 1


class Solution:
    def smallestStringWithSwaps(self, s: str, pairs: List[List[int]]) -> str:
        """
        Returns the lexicographically smallest string obtainable
        by swapping indices defined in pairs.

        Strategy:
        1. Build connected components using Union-Find.
        2. Group indices belonging to the same component.
        3. Sort characters within each component.
        4. Reassign smallest characters to smallest indices.
        """

        n = len(s)
        uf = UnionFind(n)

        # Step 1: Build connected components
        for index_a, index_b in pairs:
            uf.union(index_a, index_b)

        # Step 2: Map root -> indices in that component
        component_indices = defaultdict(list)
        for index in range(n):
            root = uf.find(index)
            component_indices[root].append(index)

        # Convert string to mutable list
        result = list(s)

        # Step 3: Sort characters within each component
        for indices in component_indices.values():

            # Collect characters
            characters = [s[i] for i in indices]

            # Sort indices and characters
            indices.sort()
            characters.sort()

            # Step 4: Place smallest characters at smallest indices
            for idx, char in zip(indices, characters):
                result[idx] = char

        return "".join(result)
```
