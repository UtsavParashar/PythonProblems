#### 685. Redundant Connection II

* https://leetcode.com/problems/redundant-connection-ii/description/

#### IMP

```python
class UnionFind:
    def __init__(self, size):
        self.parent = list(range(size + 1))
        self.rank = [0] *(size + 1)
    
    def find(self, node):
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        
        return self.parent[node]
    
    def union(self, nodea, nodeb):
        roota = self.find(nodea)
        rootb = self.find(nodeb)

        if roota == rootb:
            return False

        if self.rank[roota] < self.rank[rootb]:
            self.parent[roota] = rootb
        elif self.rank[roota] > self.rank[rootb]:
            self.parent[rootb] = roota
        else:
            self.parent[rootb] = roota
            self.rank[roota] += 1
    
        return True




class Solution:
    """
        Pattern: Union Find + Parent Tracking

        Two issues - 
        1. Two parents of a child node
        2. forms a cycle - Union Find
        3. Both 1 and 2
        
        Step 1 — Detect two-parent node
        Step 2 — Run Union-Find
        Step 3 — Decide which edge to remove

        Scenario	                    Edge to remove
        Only cycle	                    current edge
        Two parents but no cycle	    candidate2
        Two parents + cycle	            candidate1

        TC and SC - O(n)
    """
    def findRedundantDirectedConnection(self, edges: List[List[int]]) -> List[int]:

        n = len(edges)
        parents = [0] * (n+1)

        # Step1 - check if child has two parents
        candidate1 = None
        candidate2 = None
        for parent, child in edges:
            if parents[child] == 0:
                parents[child] = parent
            else:
                candidate1 = [parents[child], child] # 1st parent
                candidate2 = [parent, child] #2nd parent
                break

        # Step 2 - check if the graph has a cycle
        uf = UnionFind(n)
        for u, v in edges:
            if candidate2 and candidate2 == [u, v]:
                continue
            
            if not uf.union(u, v):
                if candidate1:
                    return candidate1 # Both cases = Two parent + cycle
                return [u, v] # only cycle found
        
        # Step 3:
        return candidate2 # only two parent

        
```
