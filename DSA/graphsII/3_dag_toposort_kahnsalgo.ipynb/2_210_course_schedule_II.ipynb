#### 210. Course Schedule II

* https://leetcode.com/problems/course-schedule-ii/description/

#### BBG

```python
from typing import List
from collections import defaultdict, deque

class Solution:
    """
        Returns a valid course order using Kahn's Algorithm (BFS Topological Sort).
        If no such ordering exists (cycle detected), returns an empty list.
        Time and Space Complexity - O(V+E)
    """
    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        # prerequisite -> list of dependent courses
        dag = defaultdict(list)

        # In-degree count: number of prerequisites for each course
        indegree = [0]*numCourses

        # Build the graph
        for dep, indep in prerequisites:
            dag[indep].append(dep)
            indegree[dep] += 1

        # Queue initialization with courses having no prerequisites
        queue = deque(course for course in range(numCourses) if indegree[course] == 0)

        res = []
        # Process courses in BFS
        while queue:
            curr_course = queue.popleft()
            res.append(curr_course)

            # Reduce in-degree for dependent courses
            for nei in dag[curr_course]:
                indegree[nei] -= 1
                if indegree[nei] == 0:
                    queue.append(nei)
        
        # Cycle detected
        if len(res) != numCourses:
            return []
        
        return res
    

# Example with cycle
Solution().findOrder(3, [[1,0],[1,2],[0,1]])
```
