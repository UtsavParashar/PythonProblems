#### 1162. As Far from Land as Possible

* https://leetcode.com/problems/as-far-from-land-as-possible/description/

```python
from collections import deque
from typing import List


class Solution:
    def maxDistance(self, grid: List[List[int]]) -> int:
        """
        Returns the maximum distance from any water cell to nearest land.
        
        Approach:
        - Multi-source BFS from all land cells
        - Each BFS layer represents distance expansion
        - Final layer depth is the answer
        """
        n: int = len(grid)
        queue: deque = deque()

        # Step 1: Initialize queue with all land cells
        for row in range(n):
            for col in range(n):
                if grid[row][col] == 1:
                    queue.append((row, col))

        # Edge cases: no land OR no water
        if not queue or len(queue) == n * n:
            return -1

        # Directions: up, down, left, right
        directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

        max_distance: int = -1

        # Step 2: Multi-source BFS
        while queue:
            level_size = len(queue)
            max_distance += 1  # Increment distance per layer

            for _ in range(level_size):
                row, col = queue.popleft()

                for dr, dc in directions:
                    new_row, new_col = row + dr, col + dc

                    # Valid unvisited water cell
                    if 0 <= new_row < n and 0 <= new_col < n and grid[new_row][new_col] == 0:
                        grid[new_row][new_col] = 1  # Mark visited
                        queue.append((new_row, new_col))

        return max_distance
```

Why max_distance starts at -1

This is subtle and often asked:

First BFS layer = land cells → distance 0
First expansion → distance 1

So:

max_distance = -1

ensures:

After processing land layer → becomes 0
Final value = correct max distance
