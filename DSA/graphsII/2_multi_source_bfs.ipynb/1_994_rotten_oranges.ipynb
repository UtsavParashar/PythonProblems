#### 994. Rotting Oranges

* https://leetcode.com/problems/rotting-oranges/description/

#### IMP

```python
from collections import deque
from typing import List

class Solution:
    """
    LeetCode 994 - Rotting Oranges

    Approach:
    Multi-source BFS to simulate the spread of rot across the grid.

    Time Complexity: O(m * n)
    Space Complexity: O(m * n) in worst case due to queue
    """
    def orangesRotting(self, grid: List[List[int]]) -> int:
        if not grid or not grid[0]:
            return 0

        rows = len(grid)
        cols = len(grid[0])

        fresh = 0
        queue = deque()
        time_elapsed = 0
        for row in range(rows):
            for col in range(cols):
                if grid[row][col] == 2:
                    queue.append((row, col))
                elif grid[row][col] == 1:
                    fresh += 1

        print(fresh)
        dirs = ((-1, 0), (1, 0), (0, -1), (0, 1))
        while queue and fresh > 0:
            for _ in range(len(queue)):
                r, c = queue.popleft()

                for dr, dc in dirs:
                    nr, nc = dr + r, dc + c
                    if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 1:
                        grid[nr][nc] = 2
                        queue.append((nr, nc))
                        fresh -= 1
        
            time_elapsed += 1
        
        print(fresh)
        return time_elapsed if fresh == 0 else -1 
```

```python
from collections import defaultdict, deque
class Solution:
    def canFinish(self, numCourses, prerequisites) -> bool:
        dag = defaultdict(set)
        for dep, indep in prerequisites:
            dag[indep].add(dep)

        queue = deque()
        for k in dag:
            queue.append(k)
        
        res = set()
        while queue:
            curr_course = queue.popleft()

            res.add(curr_course)

            for nei in dag[curr_course]:
                if nei not in res:
                    queue.append(nei)
        print(res)
        return True if len(res) == numCourses else False

        
Solution().canFinish(2, [[1, 0]])
```
