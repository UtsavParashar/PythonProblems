#### 133. Clone Graph

* https://leetcode.com/problems/clone-graph/description/

```python
from collections import deque
from typing import Optional

# TC - O(V+E)
# SC - O(V)

# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []


from typing import Optional
class Solution:
    def cloneGraph(self, node: Optional['Node']) -> Optional['Node']:
        if node is None:
            return None

        visited = {node: Node(node.val)}

        queue = deque([node])

        while queue:
            curr_node = queue.popleft()
            
            for nei in curr_node.neighbors:
                # Here we are creating the nodes if not present - One time job
                if nei not in visited:
                    visited[nei] = Node(nei.val)
                    queue.append(nei)
                # Here we are creating the edges between the nodes
                visited[curr_node].neighbors.append(visited[nei])
        
        return visited[node]
```

```python
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

def cloneGraph(node):
    if not node: return 
    visited = {} # orig_node: copied_node

    def dfs(curr):
        if curr in visited:
            return visited[curr]

        copy = Node(curr.val)
        visited[curr] = copy 

        for nei in curr.neighbors:
            copy.neighbors.append(dfs(nei))
        
        return copy

    dfs(node)
    return visited[node]


cloneGraph([])
```

❓ Why does dfs() return copy?
Because when you're building the neighbors list of the cloned node, you need the cloned version of each neighbor — and not the original.

🔎 Let's step through the key line:

copy.neighbors.append(dfs(neighbor))
Here's what's happening:

You're visiting a neighbor of the current node.

You want to append the cloned version of that neighbor to copy.neighbors.

So you call dfs(neighbor):

If that neighbor was already cloned → return the existing clone from visited.

If it wasn’t cloned → create it, recurse on its neighbors, and return the clone.

Thus:

dfs(neighbor) returns the clone of neighbor.

You append that clone to copy.neighbors.

Without returning the cloned node, you would not be able to properly link neighbors in the cloned graph.

✅ Why We Use a visited Dictionary
1. Avoid infinite loops in cycles
Graphs may contain cycles or self-loops, e.g.:

less

Edit
A -- B
|    |
C -- D
If you naively try to clone without remembering what you've already cloned, you'll loop forever, endlessly cloning the same nodes.

The visited dictionary breaks this cycle.

2. Prevent duplicating nodes
Each node should be cloned once.
If you don’t track what's been cloned:

You might clone the same node multiple times.

You'll end up with different node objects for what should be the same node.

This would violate the graph structure and result in a wrong clone.

It maps:

Keys: nodes in the original graph.

Values: their corresponding deep copies.
