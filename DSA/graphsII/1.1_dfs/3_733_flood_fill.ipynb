#### 733. Flood Fill

* https://leetcode.com/problems/flood-fill/description/



```python
# DFS is more interview friendly, until the interviewer asks for a bfs solution

from typing import List

class Solution:
    """
        TC, SC - O(m*n)
    """
    def floodFill(self, image: List[List[int]], sr: int, sc: int, color: int) -> List[List[int]]:
        rows: int = len(image)
        cols: int = len(image[0])


        original_color = image[sr][sc]

        if original_color == color:
            return image

        def dfs(r, c):
            if r < 0 or r >= rows or c < 0 or c >= cols or image[r][c] != original_color:
                return 
            
            image[r][c] = color
            
            dfs(r-1, c)
            dfs(r+1, c)
            dfs(r, c-1)
            dfs(r, c+1)

        dfs(sr, sc)
        return image
```

```python
# BFS Solution

from collections import deque
from typing import List
class Solution:
    """
    LeetCode 733 - Flood Fill

    Time Complexity: O(m * n)
        - Each cell is visited at most once.

    Space Complexity: O(m * n)
        - In worst case, queue holds entire grid.
    """
    def floodFill(self, image: List[List[int]], sr: int, sc: int, color: int) -> List[List[int]]:
        rows: int = len(image)
        cols: int = len(image[0])

        val = image[sr][sc]
        if val == color:
            return image

        image[sr][sc] = color

        queue = deque([(sr, sc)])

        while queue:
            r, c = queue.popleft()

            for dr, dc in ((-1, 0), (1, 0), (0, -1), (0, 1)):
                nr, nc = dr + r, dc + c
                if 0 <= nr < rows and 0 <= nc < cols and image[nr][nc] == val:
                    image[nr][nc] = color
                    queue.append((nr, nc))
        
        return image
        
```

```python
# DFS Solution
from typing import List
class Solution:
    def floodFill(self, image: List[List[int]], sr: int, sc: int, color: int) -> List[List[int]]:
        rows: int = len(image)
        cols: int = len(image[0])

        val = image[sr][sc]
        if val == color:
            return image


        def dfs(r, c):
            if r < 0 or r >= rows or c < 0 or c >= cols or image[r][c] != val:
                return

            image[r][c] = color

            dfs(r-1, c)
            dfs(r+1, c)
            dfs(r, c-1)
            dfs(r, c+1)

        dfs(sr, sc)
        return image
```
