#### 1514. Path with Maximum Probability

* https://leetcode.com/problems/path-with-maximum-probability/description/

```python
class Solution:
    def maxProbability(self, n: int, edges: List[List[int]], succProb: List[float], start_node: int, end_node: int) -> float:
        """
        Finds the path from start to end with maximum probability.

        Approach:
        - Build adjacency list
        - Use max heap (priority queue) to always expand the most probable path
        - Greedy Dijkstra-style traversal

        Time Complexity: O(E log V)
        Space Complexity: O(E + V)
        """
        # Step 1: Build adjacency list
        graph = defaultdict(list)
        for (src, dest), prob in zip(edges, succProb):
            graph[src].append((dest, prob))
            graph[dest].append((src, prob))

        # Step 3: Track best probability to each node
        best_prob = [0.0]*n
        best_prob[start_node] = 1.0

        # Step 2: Max heap (use negative probability for Python min-heap)
        maxheap = [(-1.0, start_node)] # prob, node


        # Step 4: Dijkstra traversal
        while maxheap:
            curr_prob, node = heapq.heappop(maxheap)
            curr_prob = -curr_prob

            # Early exit (important optimization)
            if node == end_node:
                return curr_prob

            # Skip if we already found a better path
            if curr_prob < best_prob[node]:
                continue
            
            for nei, prob in graph[node]:
                new_prob = curr_prob * prob
                # Relaxation step (maximize instead of minimize)
                if new_prob > best_prob[nei]:
                    heapq.heappush(maxheap, (-new_prob, nei))
                    best_prob[nei] = new_prob
        
        return 0.0
```
