#### 1631. Path With Minumum Effort

* https://leetcode.com/problems/path-with-minimum-effort/description/

```python
import heapq
class Solution:
    def minimumEffortPath(self, heights) -> int:
        rows, cols = len(heights), len(heights[0])
        visited = set() # (r, c)
        minheap = [(0, 0, 0)] # diff, r, c
        directions = ((-1, 0), (1, 0), (0, -1), (0, 1))

        while minheap:
            diff, r, c = heapq.heappop(minheap)

            if r==rows-1 and c==cols-1:
                return diff

            if (r, c) in visited:
                continue

            visited.add((r, c))

            for dr, dc in directions:
                nr, nc = dr+r, dc+c
                if 0<=nr<rows and 0<=nc<cols and (nr, nc) not in visited:
                    max_diff = max(diff, abs(heights[r][c] - heights[nr][nc]))
                    heapq.heappush(minheap, (max_diff, nr, nc))

Solution().minimumEffortPath([[1,2,2],[3,8,2],[5,3,5]])
```

```python
from heapq import heappush, heappop
from typing import List


class Solution:
    """
    Computes the minimum effort required to travel from the
    top-left to bottom-right cell in a height grid.

    Effort is defined as the maximum absolute height difference
    between consecutive cells along the chosen path.

    Strategy:
        Dijkstra's algorithm with a minimax relaxation rule.

    Time Complexity:
        O(m * n * log(m * n))

    Space Complexity:
        O(m * n)
    """

    def minimumEffortPath(self, heights: List[List[int]]) -> int:
        rows: int = len(heights)
        cols: int = len(heights[0])

        # Minimum effort required to reach each cell
        min_effort: List[List[int]] = [[float("inf")] * cols for _ in range(rows)]
        min_effort[0][0] = 0

        # Min heap -> (effort, row, col)
        heap: list[tuple[int, int, int]] = [(0, 0, 0)]

        directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

        while heap:
            current_effort, row, col = heappop(heap)

            # Early exit when destination reached
            if row == rows - 1 and col == cols - 1:
                return current_effort

            # Skip stale heap entries
            if current_effort > min_effort[row][col]:
                continue

            for dr, dc in directions:
                next_row = row + dr
                next_col = col + dc

                if not (0 <= next_row < rows and 0 <= next_col < cols):
                    continue

                height_diff = abs(
                    heights[row][col] - heights[next_row][next_col]
                )

                # Minimax relaxation
                new_effort = max(current_effort, height_diff)

                if new_effort < min_effort[next_row][next_col]:
                    min_effort[next_row][next_col] = new_effort
                    heappush(heap, (new_effort, next_row, next_col))

        return 0
```
