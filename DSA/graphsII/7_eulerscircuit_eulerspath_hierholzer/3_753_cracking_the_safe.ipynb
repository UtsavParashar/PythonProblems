#### 753. Cracking the Safe

* https://leetcode.com/problems/cracking-the-safe/description/

```python
# https://chatgpt.com/c/69ce1ff7-4db0-8323-bd54-cd78b1166774
class Solution:
    """
    Cracking the Safe (LeetCode 753)

    Approach:
    - Model as De Bruijn graph
    - Use Hierholzer's algorithm (DFS) to construct Eulerian circuit
    - Track visited edges instead of nodes
    """
    def crackSafe(self, n: int, k: int) -> str:
        """
            Euler's Circuit
            Based on the solutions/combination convert it to De Brujin Graph(Compressed Str)
            and return
            Eg
            if n = 2, k = 2
            possible combinations 00, 01, 11, 10
            We can compress it to 00110 and many other similar results

            Possible combinations k^n
            TC - k^n
            SC - k^n

            Node will have value (n-1) ie for n=3 and k=2, nodes will be 00, 01, 11, 10
            Edges will have combinations - 000, 010, 100, 110...  
        """

        # Edge visited set to avoid reusing same combination
        visited_edges = set()
        
        # Start node: (n-1) zeros
        start_node = '0'*(n-1)

        # Result builder (append during backtracking)
        sequence = []
        def dfs(node):
            for digit in map(str, range(k)):
                edge = node + digit
                if edge not in visited_edges:
                    visited_edges.add(edge)

                    dfs(edge[1:]) # # Move to next node (drop leftmost char)
                    
                    # Append digit AFTER recursion (post-order) - Heirholzer's algo
                    sequence.append(digit)
        
        dfs(start_node)

        # Final sequence = start_node + reversed edge traversal
        return start_node + ''.join(reversed(sequence))
        
```

```python
# Ref - https://chatgpt.com/c/68709460-6a24-800f-80c5-9a362eb9afd0

class Solution:
    def crackSafe(self, n: int, k: int) -> str:
        res = []
        start = '0'*(n-1)
        seen = set()

        def dfs(node):
            for i in map(str, range(k)):
                nei = node + i
                if nei not in seen:
                    seen.add(nei)
                    dfs(nei[1:]) # # Move to next node (drop leftmost char)
                    res.append(i)

        dfs(start)
        return start + ''.join(reversed(res))

Solution().crackSafe(n = 2, k = 2)
```
