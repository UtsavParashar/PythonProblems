#### 2097. Valid Arrangement of Pairs

* https://leetcode.com/problems/valid-arrangement-of-pairs/description/

```python
from collections import defaultdict, deque
from typing import List


class Solution:
    """
    Staff/Principal-level implementation of Eulerian Path
    using Hierholzer’s Algorithm.

    Time Complexity: O(E)
    Space Complexity: O(E)
    """

    def validArrangement(self, pairs: List[List[int]]) -> List[List[int]]:
        # Step 1: Graph + degree tracking
        adjacency: dict[int, deque[int]] = defaultdict(deque)
        in_degree: dict[int, int] = defaultdict(int)
        out_degree: dict[int, int] = defaultdict(int)

        for src, dst in pairs:
            adjacency[src].append(dst)
            out_degree[src] += 1
            in_degree[dst] += 1

        # Step 2: Find start node
        start_node = pairs[0][0]  # fallback

        for node in out_degree:
            if out_degree[node] - in_degree[node] == 1:
                start_node = node
                break

        # Step 3: Hierholzer's DFS
        path: List[int] = []

        def dfs(node: int) -> None:
            """
            Post-order DFS:
            Consume edges while traversing
            """
            while adjacency[node]:
                next_node = adjacency[node].popleft()
                dfs(next_node)

            path.append(node)

        dfs(start_node)

        # Step 4: Build edge list from node path
        path.reverse()

        result: List[List[int]] = []
        for i in range(len(path) - 1):
            result.append([path[i], path[i + 1]])

        return result
```

```python
from typing import List
class Solution:
    def validArrangement(self, pairs: List[List[int]]) -> List[List[int]]:
        # graph represents adjacency list, inOutDeg tracks in/out degree difference
        graph = defaultdict(list)
        inOutDeg = defaultdict(int)

        # Build graph and count in/out degrees
        for start, end in pairs:
            graph[start].append(end)
            inOutDeg[start] += 1  # out-degree
            inOutDeg[end] -= 1    # in-degree

        # Find starting node 
        startNode = pairs[0][0] 
        for node in inOutDeg:
            if inOutDeg[node] == 1:
                startNode = node
                break

        path = []
        def dfs(curr):
            while graph[curr]:
                nextNode = graph[curr].pop()
                dfs(nextNode)
                path.append((curr, nextNode))

        dfs(startNode)
        return path[::-1]
```
