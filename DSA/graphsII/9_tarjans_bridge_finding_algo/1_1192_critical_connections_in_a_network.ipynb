#### 1192. Critical Connections in a Network

* https://leetcode.com/problems/critical-connections-in-a-network/description/

#### The Core Logic: Tarjan’s Algorithm
* The most efficient way to solve this is using Tarjan’s Bridge-Finding Algorithm (or a similar DFS-based approach using "discovery times" and "low-link values").

##### Discovery Time (disc): The time at which a node was first visited during DFS.
##### Low-Link Value (low): The lowest discovery time reachable from the node (including itself) using back-edges in the DFS tree.
##### The Bridge Condition: An edge $(u, v)$ is a bridge if and only if $low[v] > disc[u]$. This means there is no way for $v$ or its descendants to "loop back" to $u$ or any of $u$'s ancestors without using the edge $(u, v)$.

#### Revise Tarjan's algo logic using - https://www.youtube.com/watch?v=wUgWX0nc4NY

#### Concepts to remember -
* Strong Connected Components - (Smallest Cyclic Graph)
* Low Link Values - numbering the nodes so that all nodes with same value form a SCC

```python
from typing import List, Dict, DefaultDict
from collections import defaultdict

class Solution:
    def criticalConnections(self, n: int, connections: List[List[int]]) -> List[List[int]]:
        """
        Finds all critical connections (bridges) in an undirected graph.
        
        Complexity:
            Time: O(V + E) - Each node and edge is visited once.
            Space: O(V + E) - To store the adjacency list and recursion stack.
        """

        # 1. Build Adjacency List
        # Using a defaultdict for cleaner edge management
        graph = defaultdict(list)
        for u, v in connections:
            graph[u].append(v)
            graph[v].append(u)
        
        # 2. State Management
        # disc: discovery time of each node
        # low: lowest discovery time reachable via back-edges
        # Initializing with -1 to track unvisited nodes
        discovery_time = [-1]*n
        low_links = [-1]*n
        critical_edges = []

        def find_bridges(curr_node, parent_node, current_time):
            """
            Recursive DFS to update discovery times and low-link values.
            """
            discovery_time[curr_node] = low_links[curr_node] = current_time

            for nei in graph[curr_node]:
                # If the neighbor is the parent, skip it (don't go back the way we came) CYCLE
                if nei == parent_node:
                    continue

                if discovery_time[nei] == -1:
                    # Node has not been visited, recurse
                    find_bridges(nei, curr_node, current_time+1)

                    # Upon return, update low-link of current node based on child
                    low_links[curr_node] = min(low_links[curr_node], low_links[nei])

                    # BRIDGE CONDITION:
                    # If the lowest reachable node from neighbor is still "after" 
                    # curr_node, then (curr_node, neighbor) is a bridge.
                    if low_links[nei] > discovery_time[curr_node]:
                        critical_edges.append([curr_node, nei])
                else:
                    # BACK-EDGE detected:
                    # Update low-link based on the discovery time of the visited neighbor
                    low_links[curr_node] = min(low_links[curr_node], discovery_time[nei])
            
            return current_time
        
        # Start DFS from node 0 (assuming the graph is connected as per problem constraints)
        find_bridges(curr_node=0, parent_node=-1, current_time=0)

        return critical_edges

Solution().criticalConnections(4, [[0,1],[0,2],[1,2], [1,3]])
```
