#### 451. Sort Characters By Frequency

* https://leetcode.com/problems/sort-characters-by-frequency/description/

#### BBG

```python
# use bucket sort solution over heap solution for this problem as it does not require top k 

       # Overall time complexity - O(n)
        # Overall space complexity - O(n)

from collections import Counter


class Solution:
    def frequencySort(self, s: str) -> str:
        # Edge case - empty string
        # string with single char
        # all chars are same
        # case mismatch? No - input is only lowercase
        # all chars are different

        # Solution - BUCKET SORT - O(n) solution
        # Step 1 - Count the freq of each char using Counter - O(n)
        # Step 2 - using list of list of default_dict(list), place all chars against the freq to form the bucket
        # list of list - [[], ['r', 't'],['e'], [], []] - we are using the logic that the total number of lists in the list can't be greater than n -- 0<len(list)<n+1
        # why? because max could happen - is that all elements are same in input 'aaa' then list of list [[], [], [], ['a']] 
        # {2: ['e'], 1: ['r', 't']} -- slightly more space with defaultdict
        # Step 3 - traverse through the list of list from last to start as we want the output in descending order (most frequent char first)

        # Overall time complexity - O(n)
        # Overall space complexity - O(n)

        # Follow up question - can we return the deteministic output in lexicographic order by keeping the time complexity same as O(n)
        # eg - for tree always returns the output 'eert' and not probablistic output 'eert'|'eetr'
        # Yes - we can sort the bucket, and since sorting the elements(max 256 - all chars included lower/upper/special char) will have time complexity nearing O(1) hence it won't affect the overall time complexity.

        n = len(s)
        # step 1
        freq = Counter(s)

        # step 2
        # bucket = defaultdict(list) could be used which would take slightly more space because of hm
        # list of list with keep is in order of bucket sort but will have some empty element(inner list)
        bucket = [[] for _ in range(n+1)] # n included
        for ch, count in freq.items():
            bucket[count].append(ch)

        # step 3
        # no element can occur 0 times as filtered in Counter hence 0 is not inclusive
        res = [] # using list as immutable string would create a new object everytime
        for count in range(n, 0, -1):
            if bucket[count]:
                bucket[count].sort() # deterministic - lexicographic output
                for ch in bucket[count]:
                    res.append(ch * count)

        return ''.join(res)

Solution().frequencySort('aaaa')
```

```python
# Heap Solution - Not preferred for this problem because of O(N log K) TC

import heapq

class Solution:
    def frequencySort(self, s: str) -> str:
        n = len(s)
        freq_counter = Counter(s)

        max_heap = []
        for char, freq in freq_counter.items():
            heapq.heappush(max_heap, (-freq, char))

        res = []
        while max_heap:
            res.append(heapq.heappop(max_heap))

        return ''.join(char * -freq for freq, char in res)
```
