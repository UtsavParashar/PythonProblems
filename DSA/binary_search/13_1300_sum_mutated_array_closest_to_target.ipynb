#### 1300. Sum of Mutated Array Closest to Target

* https://leetcode.com/problems/sum-of-mutated-array-closest-to-target/description/

#### Bloomberg

```python
# TC - O(n log M) where M is max of array 
# SC - O(1)
# https://chatgpt.com/c/694fd5a0-7ae4-8323-b44f-e74a002b1696

from typing import List

class Solution:
    def findBestValue(self, arr: List[int], target: int) -> int:
        # Helper function:
        # For a given value v, compute the sum of the mutated array
        # where every element greater than v is replaced by v
        def mutated_sum(v: int) -> int:
            return sum(min(x, v) for x in arr)

        # Binary search boundaries:
        # v can range from 0 to max element in the array
        left, right = 0, max(arr)

        # best_val stores the answer closest to target found so far
        best_val = 0

        # min_diff stores the minimum absolute difference seen so far
        min_diff = float('inf')

        # Binary search on the value v
        while left <= right:
            mid = (left + right) // 2

            # Compute the mutated sum for current mid
            curr_sum = mutated_sum(mid)

            # Difference between current sum and target
            diff = abs(curr_sum - target)

            # Update best_val if:
            # 1) current difference is smaller, OR
            # 2) difference is same but mid is smaller (tie-breaking rule)
            if diff < min_diff or (diff == min_diff and mid < best_val):
                min_diff = diff
                best_val = mid

            # If mutated sum is less than target,
            # we need a larger v to increase the sum
            if curr_sum < target:
                left = mid + 1
            else:
                # If mutated sum is greater than or equal to target,
                # reduce v to try to get closer
                right = mid - 1

        # Return the value v that produces the closest mutated sum
        return best_val
    
Solution().findBestValue([4, 9, 3], 10)
Solution().findBestValue([2, 5, 3], 10)
```
