#### 154. Find Minimum in Rotated Sorted Array II

* https://leetcode.com/problems/find-minimum-in-rotated-sorted-array-ii/description/

#### BBG - IMP

```python
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        if not nums:
            return 

        n: int = len(nums)
        left: int = 0
        right: int = n - 1

        if nums[left] < nums[right]:
            return nums[left]

        while left < right:
            mid = left + (right - left)//2
            # Case 1: Minimum is strictly in the right half
            if nums[mid] > nums[right]:
                left = mid + 1
            # Case 2: Minimum is in the left half (including mid)
            elif nums[mid] < nums[right]:
                right = mid
            # Case 3: nums[mid] == nums[right]
            # Cannot determine which side contains the minimum
            # Safe move: shrink the right boundary
            else:
                right -= 1

        return nums[left]

        
```
