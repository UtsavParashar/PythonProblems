```python
# https://chatgpt.com/c/69575752-53e0-8321-81ba-09596caa2e7c
# Two approaches
# 1. StairCase approach - TC - O(m+n), SC - O(1) - Preferred Approach
# Start and top right, if target is small move left if target is big move down
# “Staircase search only works if both rows and columns are monotonic(strictly increasing or decreasing). If either constraint is violated, the pruning logic breaks.”


from typing import List

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        # Two approaches
        # 1. StairCase approach - TC - O(m+n), SC - O(1) - Preferred Approach
        # Start and top right, if target is small move left if target is big move down
        # why this approach works - Unlike lc 74 this grid is not sorted as a whole
        # rather each row is sorted and each col is sorted
        # Starting from any other corner of the grid would not work
        # application - risk metrics with time on x axis 
        # and increasing risk for different tickers on y axis

        # 2. run Binary search on each row - TC - O(n log m), SC - O(1)
        # traverse through each row of the grid and binary search for the target

        if not matrix:
            return False

        rows, cols = len(matrix), len(matrix[0])
        r, c = 0, cols-1

        while r < rows and c >= 0:
            if matrix[r][c] == target:
                return True
            elif matrix[r][c] > target: # curr val less than target - move left
                c -= 1
            else: # move down
                r += 1
        return False

Solution().searchMatrix([[1,4,7,11,15],[2,5,8,12,19],[3,6,9,16,22],[10,13,14,17,24],[18,21,23,26,30]], target = 5)
```

```python
# 2. run Binary search on each row - TC - O(n log m), SC - O(1)
# This solution would be preferrable when either row is not sorted or col is not sorted
# Eg - rows sorted only - Across instruments: No consistent ordering ❌ and Prices sorted by time ✔
# Rows    → Instruments, Columns → Time buckets

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        # Second approach - Binary Search with Range Pruning
        # 2. run Binary search on each row - TC - O(n log m), SC - O(1)
        # traverse through each row of the grid and binary search for the target
        if not matrix:
            return False

        for row in matrix:
            l, r = 0, len(row)-1
            # range pruning -  TC is O(n log m) but in practise its faster
            if target < row[l] or target > row[r]:
                continue

            while l <= r:
                m = (l+r)//2
                if row[m] == target:
                    return True
                elif row[m] < target:
                    l = m + 1
                else:
                    r = m - 1

        return False

        
```

```python
# 2. run Binary search on each row - TC - O(n log m), SC - O(1)

import bisect

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        # Second approach
        # 2. run Binary search on each row - TC - O(n log m), SC - O(1)
        # traverse through each row of the grid and binary search for the target
        if not matrix:
            return False

        for row in matrix:
            if target < row[0] or target > row[len(row)-1]:
                continue
            idx = bisect.bisect_left(row, target)
            if idx < len(row) and row[idx] == target:
                return True
        return False

Solution().searchMatrix([[1,4,7,11,15],[2,5,8,12,19],[3,6,9,16,22],[10,13,14,17,24],[18,21,23,26,30]], target = 5)
```

```python
# What is rows are not sorted and only column is sorted - then Go with Binary search

def searchMatrix(matrix, target):
    rows, cols = len(matrix), len(matrix[0])

    for c in range(cols):
        if target < matrix[0][c] or target > matrix[rows-1][c]:
            continue
        lo, hi = 0, rows - 1
        while lo <= hi:
            mid = (lo + hi) // 2
            if matrix[mid][c] == target:
                return True
            elif matrix[mid][c] < target:
                lo = mid + 1
            else:
                hi = mid - 1

    return False
```
