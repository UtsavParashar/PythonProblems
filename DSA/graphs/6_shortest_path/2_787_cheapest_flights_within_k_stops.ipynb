### 787. Cheapest Flights Within K Stops

* https://leetcode.com/problems/cheapest-flights-within-k-stops/

```python
# Djikshtra's algo

import heapq
from collections import defaultdict
from typing import List

class Solution:
    """
        Dijkstra-style BFS with stop constraint.

    Time Complexity:
        O((E + V) log V) in practical scenarios.
        Worst case: O(E log E)

    Space Complexity:
        O(V + E)
    """
    def findCheapestPrice(self, n: int, flights: List[List[int]], src: int, dst: int, k: int) -> int:
        graph = defaultdict(list)

        for src_, dest, cost in flights:
            graph[src_].append((dest, cost))

        # Min-heap: (total_cost, current_city, stops_used)
        minheap = [(0, src, 0)] # cost, src, stops_used

        # Dictionary to track minimum stops used to reach a city
        # Helps prune unnecessary states
        visited = {}

        while minheap:
            curr_cost, curr_city, stops_used = heapq.heappop(minheap)

            # If destination reached within allowed stops
            if curr_city == dst:
                return curr_cost

            # If stops exceed allowed limit, skip
            if stops_used > k:
                continue

            # Pruning:
            # If we've already visited this city with fewer stops, skip
            if curr_city in visited and visited[curr_city] <= stops_used:
                continue

            visited[curr_city] = stops_used

            for nei_city, cost in graph[curr_city]:
                heapq.heappush(minheap, (curr_cost + cost, nei_city, stops_used+1))
            
        return -1
```

```python
# Ref - https://www.youtube.com/watch?v=5eIK3zUdYmE - Explanation is distastrous
# Bellman-Ford Algo

class Solution:
    def findCheapestPrice(self, n: int, flights, src, dst, k) -> int:
        INF = float('inf')
        prices = [INF]*n
        prices[src] = 0

        for _ in range(k+1):
            tmp = prices.copy()
            for u, v, cost in flights:
                if prices[u] == INF:
                    continue
                if prices[u]+cost < tmp[v]:
                    tmp[v]= prices[u]+cost
            prices = tmp

        return -1 if prices[dst]==INF else prices[dst]


Solution().findCheapestPrice(n = 4, flights = [[0,1,100],[1,2,100],[2,0,100],[1,3,600],[2,3,200]], src = 0, dst = 3, k = 1)
```
