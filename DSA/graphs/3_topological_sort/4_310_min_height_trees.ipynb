#### 310. Min Height Trees

* https://leetcode.com/problems/minimum-height-trees/description/

```python
from collections import defaultdict

class Solution:
    def findMinHeightTrees(self, n, edges):
        if n==1:
            return [0]

        graph = defaultdict(set)

        for i, nei in edges:
            graph[i].add(nei)
            graph[nei].add(i)

        leaf_nodes = [node for node in graph if len(graph[node])==1]

        remaining_nodes = n
        

        while remaining_nodes > 2:
            remaining_nodes -= len(leaf_nodes)
            new_leaves = []

            for leaf in leaf_nodes:
                nei = graph[leaf].pop()
                graph[nei].remove(leaf) 

                if len(graph[nei]) == 1:
                    new_leaves.append(nei)

            leaf_nodes = new_leaves

        return leaf_nodes


Solution().findMinHeightTrees(n = 4, edges = [[1,0],[1,2],[1,3]])
Solution().findMinHeightTrees(n = 6, edges = [[3,0],[3,1],[3,2],[3,4],[5,4]])
```

Why at most 2 centroids?
A centroid in a tree is defined as:

A node that, when chosen as the root, results in the minimum possible height of the tree.

For any tree:

✅ There can be 1 or 2 centroids, never more.

Why?
Think of the tree as being "peeled" layer by layer:

You remove all leaves in each round.

After each round, the inner structure of the tree is revealed.

Eventually, you are left with:

1 node if the tree has an odd number of layers — single centroid.

2 adjacent nodes if the tree has an even number of layers — both are centroids.

Graph Theory Intuition
A tree is an acyclic connected graph with n nodes and n - 1 edges.

In a tree:

There is exactly 1 unique simple path between any two nodes.

The longest such path is called the diameter of the tree.

The centroids lie:

At the middle point(s) of this diameter.

If the diameter is odd-length → 1 centroid (middle node).

If even-length → 2 centroids (the two central nodes).
