#Intersection of arrays
* Given two arrays, return the intersection of two lists. (both cases - when sorted and when not)

// Example
- A = [1, 2, 3, 4]
- B = [2, 4, 6]
- Result - [2, 4]

#### All solutions below have TC O(m + n), Sorted array SC - O(1), Unsorted array  SC - O(n)

```python
# def intersection_(A, B):
#     return set(A).intersection(set(B))

# when arrays are sorted
def intersection_(a, b):
    i = 0
    j = 0

    lena = len(a)
    lenb = len(b)

    res = []
    while i < lena and j < lenb:
        if a[i] == b[j]:
            res.append(a[i])
            i += 1
            j += 1
            while i < lena and a[i] == a[i-1]:
                i += 1
            while j < lenb and b[j] == b[j-1]:
                j += 1
        elif a[i] < b[j]:
            i += 1
        else:
            j += 1
    
    return res

intersection_([1, 1, 2, 3, 4], [1, 1, 2, 4, 6])
```

```python
# when arrays are not sorted and duplicates are allowed

from collections import Counter
def intersection_(a, b):
    counter = Counter(a)
    res = []
    for num in b:
        if num in counter:
            res.append(num)
            counter[num] -= 1
    
    return res

intersection_([1, 3, 2,4, 1], [1, 4, 2, 1, 6])
```

```python
# when arrays are not sorted and duplicates are not allowed

from collections import Counter
def intersection_(a, b):
    seen = {num: True for num in a}
    res = []
    for num in b:
        if num in seen:
            res.append(num)
            del seen[num]
    
    return res

intersection_([1, 3, 2,4, 1], [1, 4, 2, 1, 6])
```
