#### 2593. Find Score of an Array After Marking All Elements
* https://leetcode.com/problems/find-score-of-an-array-after-marking-all-elements/description/

```python
class Solution:
    def findScore(self, nums: List[int]) -> int:
        n = len(nums)
        is_marked = [False] * n

        res = 0
        for i, v in sorted(enumerate(nums), key=lambda x: x[1]):
            if is_marked[i]:
                continue
            
            res += v
            is_marked[i] = True

            if i > 0:
                is_marked[i-1] = True
            
            if i < n-1:
                is_marked[i+1] = True
        
        return res
```
