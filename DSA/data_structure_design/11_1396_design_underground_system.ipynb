#### 1396. Design Underground System
* https://leetcode.com/problems/design-underground-system/description/
### BBG - IMP

```python
class UndergroundSystem:

    def __init__(self):
        self.route_stats = {} # (startStation, endStation) -> (total_time, trip_count)
        self.active_users = {} # user_id -> (start_station, starttime)


    def checkIn(self, id: int, stationName: str, t: int) -> None:
        self.active_users[id] = (stationName, t)

    def checkOut(self, id: int, stationName: str, t: int) -> None:
        start_station, start_time = self.active_users.pop(id)
        trip_time = t - start_time

        route_key = (start_station, stationName)
        if route_key not in self._route_stats:
            self._route_stats[route_key] = (0, 0)

        existing_time, existing_trips = self.route_stats[route_key]
        self.route_stats[route_key] = (existing_time + trip_time, existing_trips + 1)


    def getAverageTime(self, startStation: str, endStation: str) -> float:
        total_time, trip_count = self.route_stats[(startStation, endStation)]
        return total_time/trip_count
```
