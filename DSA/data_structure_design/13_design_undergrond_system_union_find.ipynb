## BBG 


```python
from collections import defaultdict, deque

class StationMap:
    def __init__(self, station_dict):
        self.station_dict = station_dict
        self.graph = defaultdict(set)
        for src, dest_list in self.station_dict.items():
            for dest in dest_list:
                self.graph[src].add(dest)
                self.graph[dest].add(src)

        self.reachable_stations = defaultdict(set)
        for src in self.graph:
            for dest in self.graph:
                if self._is_reachable(src, dest):
                    self.reachable_stations[src].add(dest)

    def _is_reachable(self, src, dest):
        if src not in self.graph or dest not in self.graph:
            return False
        
        queue = deque([(src)])
        visited = {src}

        while queue:
            curr_station = queue.popleft()

            if curr_station == dest:
                return True
            
            for next_station in self.graph[curr_station]:
                if next_station not in visited:
                    visited.add(next_station)
                    queue.append(next_station)
        
        return False
    

    def is_reachable(self, src, dest):
        if src not in self.graph or dest not in self.graph:
            return False
        
        return True if dest in self.reachable_stations[src] else False
    



```
