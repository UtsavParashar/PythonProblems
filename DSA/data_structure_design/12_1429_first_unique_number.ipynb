#### First Unique Number

* https://neetcode.io/problems/first-unique-number/question

#### BBG

```python
from collections import OrderedDict
from typing import List


class FirstUnique:
    """
    Maintains a stream of integers and supports:
    - O(1) amortized insertion
    - O(1) retrieval of the first unique element
    """

    def __init__(self, nums: List[int]):
        # Frequency map to track occurrence count
        self.frequency = {} # num -> freq

        # OrderedDict to maintain insertion order of unique elements
        self.unique_order = OrderedDict() # num -> None, store only the elements which have occured once

        for number in nums:
            self.add(number)

    def showFirstUnique(self) -> int:
        """
        Returns the first unique number in the stream.
        Time Complexity: O(1)
        """
        if not self.unique_order:
            return -1

        # next(iter(...)) gets first key in OrderedDict
        return next(iter(self.unique_order))

    def add(self, value: int) -> None:
        """
        Adds a value to the stream.
        Time Complexity: O(1) amortized
        """
        count = self.frequency.get(value, 0)
        self.frequency[value] = count + 1

        if count == 0:
            # First time seeing this value → candidate unique
            self.unique_order[value] = None
        elif count == 1:
            # Value is no longer unique → remove from OrderedDict
            self.unique_order.pop(value, None)
```
