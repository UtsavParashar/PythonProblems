### 90. Subsets II
* https://leetcode.com/problems/subsets-ii/description/

### ['Bloomberg', 'Meta', 'Amazon', 'Google', 'Apple']

```python
# IF we keep the solution same as leetcode 78, then it will result in duplicates in the res
# TC - 2^n * n # This is the standard upper bound for subsetâ€“generation problems.
# SC - O(n) # O(n) (recursion depth + temporary subset)

class Solution:
    def subsetsWithDup(self, nums):
        # Edge case - Unsorted array
        # we will have to sort it to manage duplicate elements

        if not nums:
            return

        nums.sort()

        res, sol = [], []
        n = len(nums) 
        def bt(i):
            res.append(sol[:])

            for j in range(i, n):
                if j > i and nums[j] == nums[j-1]:
                    continue

                sol.append(nums[j])
                bt(j+1)
                sol.pop()
                

        bt(0)
        return res
```
