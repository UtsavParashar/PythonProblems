#### 77. Combinations

* https://leetcode.com/problems/combinations/description/

#### ['Bloomberg', 'Meta', 'Amazon', 'Google', 'Apple']

```python
class Solution:
    def combine(self, n: int, k: int):
        # Edge cases
        # 1. n=0 and/or k = 0
        # 2. k > n
        # 3. k = 1, return 1..n
        # [1,2] == [2, 1]

        res, sol = [], []

        def bt(i):
            # Pruning: if remaining numbers < remaining picks, stop early
            if len(sol) + (n-i+1) < k:
                return

            # If the combination is ready
            if len(sol) == k:
                res.append(sol[:])
                return

            for j in range(i, n+1):
                sol.append(j)
                bt(j+1)
                sol.pop()

        bt(1)
        return res

Solution().combine(4, 3)
```

```python
class Solution:
    def search(self, nums, target: int) -> bool:
        if not nums:
            return False
        
        l, r = 0, len(nums)-1

        while l <= r:
            m = (l+r)//2

            if nums[m] == target:
                return True
            elif nums[l] < nums[m]:
                if nums[l] <= target < nums[m]:
                    r = m - 1
                else:
                    l = m + 1
            else:
                if nums[m] < target <= nums[r]:
                    l = m + 1
                else:
                    r = m - 1

        return False
    
Solution().search([1,0,1,1,1], 0)
```
