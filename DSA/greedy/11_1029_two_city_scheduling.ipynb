#### 1029. Two City Scheduling

* https://leetcode.com/problems/two-city-scheduling/description/

#### BBG - IMP

```python
class Solution:
    """
    Greedy assignment is often used when capacity is fixed.
    For streams suggest heaps
        
        This is one of those problems which could be sorted in multiple way.
        We would prefer below greedy solution because of nlogn TC and O(1)SC

        
        Below Greedy Solution -
        We leverage the fact the getting the diff between C[A]-Cost[B] and sorting them would keep the first half where tickets of A is less than B and the second half will have cheap tickets of B.
        Tc - O(n log n)
        SC - O(1)

        In greedy solution also you can use space of O(n) but let's not do that.

        2. Heap Based solution
        TC - O(n log n)
        SC - O(n)

        3. DP Solution
        TC - O(n^2)
        SC - O(n)

        4. BT Solution
        TC - O(exponential)

    """
    def twoCitySchedCost(self, costs: List[List[int]]) -> int:
        """
        Greedy solution for Two City Scheduling.

        Strategy:
        Sort people by cost difference (A - B).
        Send first half to A and second half to B.
        """
        if not costs:
            return 0

        # Sort by relative savings of choosing A over B
        costs.sort(key=lambda c: c[0] - c[1])

        total_people = len(costs)
        half = total_people//2

        res = 0
        for i in range(half):
            res += costs[i][0]

        for i in range(half, total_people):
            res += costs[i][1]

        return res
```
