```python
## Mutable Solution - by reusing the same matrix for prefix sum
## Logic defined beautifully within first 5 mins -  https://www.youtube.com/watch?v=KE8MQuwE2yA

from typing import List

class NumMatrix:
    """
        3 Approaches
    
        Tradeoff - discuss the short comings in terms of mutability
        and space between 1st and 2nd approach and then 
        use one of the 3 approaches accordingly

        1. Apply presum on the input i.e update existing matrix - MUTABLE SOLUTION
            TC - O(m*n) preprocessing, O(1) query
            SC - O(1)
            Original data is lost - mutability issue

        2. Create a new matrix with prefix sum - IMMUTABLE SOLUTION (As asked in the question)
            TC - O(m*n) preprocessing, O(1) query
            SC - O(m+n)
            Original data remains intact - immutable

    """
    def __init__(self, matrix: List[List[int]]):
        self.matrix = matrix
        rows, cols = len(matrix), len(matrix[0])

        # row wise prefix
        for r in range(rows):
            for c in range(1, cols):
                self.matrix[r][c] += self.matrix[r][c-1]

        # column wise prefix
        for r in range(1, rows):
            for c in range(cols):
                self.matrix[r][c] += self.matrix[r-1][c]

    def sumRegion(self, row1: int, col1: int, row2: int, col2: int) -> int:
        ''' Below Brute Force O(n^2) solution raises TLE
        res = 0
        for r in range(row1, row2+1):
            for c in range(col1, col2+1):
                res += self._num_matrix[r][c]
        return res
        '''
        total = self.matrix[row2][col2]

        # Logic defined beautifully here 
        # https://www.youtube.com/watch?v=KE8MQuwE2yA
        if row1 > 0:
            total -= self.matrix[row1-1][col2]
        if col1 > 0:
            total -= self.matrix[row2][col1-1]
        if row1>0 and col1>0:
            total += self.matrix[row1-1][col1-1]
        
        return total


# Your NumMatrix object will be instantiated and called as such:
# obj = NumMatrix(matrix)
# param_1 = obj.sumRegion(row1,col1,row2,col2)
```
