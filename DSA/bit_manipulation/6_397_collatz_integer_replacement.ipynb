#### 397. Integer Replacement
* https://leetcode.com/problems/integer-replacement/description/
#### Collatz Eqn
#### IMP BBG

```python
# Preferred solution - Iterative + Greedy as counting traling zero gives exact greedy
# TC - O(log n)
# SC - O(1)
# Deep discussion - https://chatgpt.com/c/6954e962-3cbc-8321-85ca-b836a9c3bc3e

class Solution:
    def integerReplacement(self, n: int) -> int:
        # This is collatz seq which tells that using these eqns every number would reach 1 eventually - np complete
        # using bit manipulation
        # right shift operator == n//2
        # we want to use the fact that 
        # if the number of traling zeros are more the number of even number of computations are that many
        # eg 8 in binary is 1000 - 3 divide by 2s to reach 1 - 8/2, 4/2, 2/2
        # eg 12 in binary is 1100 -  3 divide by 2s to reach odd - 12/2, 6/2

        # for odd numbers - we want to check how many traling zeros are there after AND with 2
        # we want to check last binary digit and 'AND'(& - BIT MASK) it with 2
        # if it returns 0 then n-1 would require less divisions with an exception to 3
        # if it is not 0 then n+1 would require less divisions
        # Eg 5 binary is 101 - last 2 bit 01 AND 10(2) = 00 then n-1
        # Eg 7 binary is 111 - last 2 bit 11 AND 10 = 10 then n+1

        steps = 0
        while n > 1: # or n != 1
            if n % 2 == 0:
                n >>= 1
            else:
                if n == 3 or (n&2) == 0:
                    n -= 1
                else:
                    n += 1
            steps += 1
        return steps
        
```

```python
# Memoization solution - Not preferred because
# Recursion stack - Decision tree explodes
# TC - O(n) - worst case
# SC - O(n)

class Solution:
    def integerReplacement(self, n: int) -> int:
        memo = {1: 0}

        def power(x):
            if not x in memo:
                if x % 2 == 0:
                    memo[x] = 1 + power(x//2)
                else:
                    memo[x] = 1 + min(power(x-1), power(x+1))
            return memo[x]

        return power(n)
        
```
