```python
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        """
            Three Solutions
            1. Using XOR - O(n), O(1)
            since a^a = 0 and a^0 = a
            we will use this to find the missing using the relation
            missing = n
            missing ^= index ^ num
            remember the example [0, 1, 2] 
            # answer will be 3, missing will alway be 3

            2. Basic Maths - O(n), O(1)
            expected sum = n * (n + 1) // 2
            current sum = sum(nums)
            res = expected_sum - current_sum

            3. Using set - O(n), O(n)
            add all elements from 0 to n in a set
            traverse through nums and remove the num
            the element left in the set is the ans
        """

        n = len(nums)
        nums_set = set(range(n+1))

        for num in nums:
            nums_set.remove(num)

        return nums_set.pop()

        
```
