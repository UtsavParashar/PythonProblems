#### 1387. Sort Integers by The Power Value

* https://leetcode.com/problems/sort-integers-by-the-power-value/description/

#### Collatz Seq
#### IMP - BBG

```python
# Iterative + Memoization - Preferred
# TC - O(n log n)
# SC - O(n) # amortized - memo

class Solution:
    def getKth(self, lo: int, hi: int, k: int) -> int:
        # Memoization dict: number -> power value
        memo = {1: 0}

        def power(x):
            orig = x 
            steps = 0
    
            while x > 1:
                if x in memo:
                    steps += memo[x]
                    break

                if x % 2 == 0:
                    x >>= 1
                else:
                    x = 3 * x + 1
                steps += 1
            
            # Not storing steps for each value of intermediate step
            # this keeps memory usage deterministic
            # cache could be used
            # affects speed a bit
            memo[orig] = steps
            print(memo)
            return steps

        arr = [(power(i), i) for i in range(lo, hi+1)]
        arr.sort()
        return arr[k-1][1]

Solution().getKth(7, 11, 4)
```

```python
# Recursion + Memoization - Not preferred
# Stack overflow
# storing all intermediate seq
# TC - O(n log n) # O(N log M) (amortized) + O(N log N) for sorting
# since sorting is the dominant factor here, so TC is O(nlogn)
# SC - O(M + N)


class Solution:
    def getKth(self, lo: int, hi: int, k: int) -> int:
        # Memoization dict: number -> power value
        memo = {1: 0}

        def power(x):
            if x not in memo:
                if x % 2 == 0:
                    memo[x] = 1 + power(x//2)
                else:
                    memo[x] = 1 + power(3*x+1)
            print(memo)
            return memo[x]

        arr = [(power(i), i) for i in range(lo, hi+1)]
        arr.sort()
        return arr[k-1][1]

Solution().getKth(7, 11, 4)    
```
