##### 390. Elimination Game

* https://leetcode.com/problems/elimination-game/description/

#### BBG - IMP

```python
# https://chatgpt.com/c/6982c717-28f8-83a2-9adb-8902b8f97b41

class Solution:
    """
    Elimination Game
    Time Complexity: O(log n)
    Space Complexity: O(1)

    Remember ltr == not ltr and not -ltr as -1 is also True then head will increment with each iteration
    """

    def lastRemaining(self, n: int) -> int:
        # First remaining number in the sequence
        head = 1
        
        # Distance between consecutive remaining numbers
        step = 1
        
        # Count of remaining numbers
        remaining = n
        
        # Direction flag
        left_to_right = True

        while remaining > 1:
            # If we eliminate from left,
            # or from right with odd remaining count,
            # the head must move forward
            if left_to_right or remaining % 2 == 1:
                head += step

            # After each round:
            # - half numbers are eliminated
            # - step doubles
            remaining //= 2
            step *= 2
            left_to_right = not left_to_right

        return head
```
