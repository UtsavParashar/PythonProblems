#### 978. Longest Turbulent Subarray

* https://leetcode.com/problems/longest-turbulent-subarray/description/

### BBG - IMP

```python
## https://chatgpt.com/c/69818238-d2b4-8322-925e-5fa654ce6321

class Solution:
    def maxTurbulenceSize(self, arr: list[int]) -> int:
        """
        Returns the length of the longest turbulent subarray.

        A turbulent subarray alternates between greater-than and less-than
        comparisons between consecutive elements.

        Time Complexity: O(n)
        Space Complexity: O(1)

        Suitable for streaming / real-time data processing.
        """

        n = len(arr)
        if n < 2:
            return n

        max_length = 1
        current_length = 1

        # Previous comparison result:
        #  1  -> arr[i-1] < arr[i]
        # -1  -> arr[i-1] > arr[i]
        #  0  -> equal
        prev_comparison = 0

        for i in range(1, n):
            if arr[i] > arr[i - 1]:
                current_comparison = 1
            elif arr[i] < arr[i - 1]:
                current_comparison = -1
            else:
                current_comparison = 0

            if current_comparison == 0:
                # Equal elements break turbulence
                current_length = 1
            elif current_comparison == -prev_comparison:
                # Alternation continues
                current_length += 1 # Streak Continues
            else:
                # Start a new turbulent window
                current_length = 2 # Streak ends

            prev_comparison = current_comparison
            max_length = max(max_length, current_length)

        return max_length
```
