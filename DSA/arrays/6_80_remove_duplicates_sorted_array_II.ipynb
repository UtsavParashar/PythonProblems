#### 80. Remove Duplicates from Sorted Array II
* https://leetcode.com/problems/remove-duplicates-from-sorted-array-ii/description/

#### ['Bloomberg', 'Meta', 'Amazon', 'Apple', 'Uber']

```python
class Solution:
    def removeDuplicates(self, nums) -> int:
        n = len(nums)
        if not nums or n < 2:
            return n

        l = 2
        for r in range(2, n):
            if nums[r] > nums[l-2]:
                nums[l] = nums[r]
                l += 1
        
        return l
        
```

```python
## TC - O(n)
## SC - O(1)
## Generic solution - could be reused for leetcode 80 - Remove Duplicates from Sorted Array I
## https://leetcode.com/problems/remove-duplicates-from-sorted-array-ii/description/

class Solution:
    def removeDuplicates(self, nums) -> int:
        # Edge cases
        # No element
        # All elements with same value

        if not nums:
            return

        i = 0
        for n in nums:
            if i < 2 or n > nums[i-2]:
                nums[i] = n
                i += 1

        return i
```
