#### 1169. Invalid Transactions

* https://leetcode.com/problems/invalid-transactions/description/

```python
class Solution:
    def invalidTransactions(self, transactions: List[str]) -> List[str]:
        # step 1 - place the transactions with index in a list of tuples 
        # [(name, time, amt, city, idx), ()...]
        # why storing indexes - to segregate the invalid transactions in next steps 
        # Time Complexity for this step - O(n)
        
        # step 2 - group the transactions by name, using defaultdict 
        # {name: (name, time, amt, city, idx)}
        # Time Complexity for this step - O(n)

        # step 3 - create an invalid_set set, to store invalid transaction indexes

        # Time Complexity for below steps - O(n log n)
        # step 4 - traverse through the default_dict for all transactions within a name
        # step 5 - sort transactions within a name using time for sorting invalid transactions by time
        # step 6 - apply rule 1 -> place invalid index in set
        # step 7 - for rule 2, check if previous transactions till 0th tranctions are within 60 mins. If yes, check if city is different, 
        # if yes - mark the current index and the previous transaction index as invalid
        # step 8 - filter invalid transactions from input using invalid_set using list comprehension and return it 

        # Overall TC - O(n) + O(n) + O(nlogn) ~ O(n log n)
        # SC - O(n) + O(n) ~ O(n) 

        # Solution
        # step 1 - place the transactions with index in a list of tuples 
        # [(name, time, amt, city, idx), ()...]

        AMT_LIMIT = 1000
        TIME_LIMIT = 60 # inclusive

        parsed = []
        for idx, txn in enumerate(transactions):
            name, time, amt, city = txn.split(',')
            parsed.append((name, int(time), int(amt), city, idx)) 
        print(parsed)
        
        # step 2 - group the transactions by name, using defaultdict 
        # {name: (name, time, amt, city, idx)}
        # Time Complexity for this step - O(n)

        by_name = defaultdict(list)
        for txn in parsed:
            by_name[txn[0]].append(txn)
        print(by_name)

        # step 3 - create an invalid_set set, to store invalid transaction indexes
        invalid_set = set()

        # step 4 - traverse through the default_dict for all transactions within a name
        # step 5 - sort transactions within a name using time for sorting invalid transactions by time
        # step 6 - apply rule 1 -> place invalid index in set
        # step 7 - for rule 2, check if previous transactions till 0th tranctions are within 60 mins. If yes, check if city is different, 
        # if yes - mark the current index and the previous transaction index as invalid

        for name, txns in by_name.items():
            txns.sort(key = lambda x: x[1]) # x[1] is time

            n = len(txns)
            for i in range(n):
                name_i, time_i, amt_i, city_i, idx_i = txns[i]

                # rule 1
                if amt_i > AMT_LIMIT:
                    invalid_set.add(idx_i)
                
                # rule 2 - use sorted time transactions to compare all transactions backwards with i within timelimit 
                j = i - 1
                while j >= 0 and time_i - txns[j][1] <= TIME_LIMIT:
                    if city_i != txns[j][3]:
                        invalid_set.add(idx_i)
                        invalid_set.add(txns[j][4])
                    j -= 1
        
        # step 8 - filter invalid transactions from input using invalid_set using list comprehension and return it
        return [transactions[i] for i in invalid_set]

```
