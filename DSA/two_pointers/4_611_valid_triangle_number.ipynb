#### 611. Valid Triangle Number

* https://leetcode.com/problems/valid-triangle-number/description/

#### BBG - IMP

```python
from typing import List

class Solution:
    def triangleNumber(self, nums: List[int]) -> int:
        """
        Counts the number of valid triangles using sorting + two pointers.

        Time Complexity: O(n^2)
        Space Complexity: O(1) (in-place sort)
        """

        nums.sort()
        n = len(nums)
        valid_triangle_count = 0

        # Fix the largest side one by one
        for largest_index in range(n - 1, 1, -1):
            left = 0
            right = largest_index - 1

            while left < right:
                if nums[left] + nums[right] > nums[largest_index]:
                    # All values between left and right form valid triangles
                    valid_triangle_count += (right - left)
                    right -= 1
                else:
                    left += 1

        return valid_triangle_count
```
