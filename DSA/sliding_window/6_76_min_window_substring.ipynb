#### 76. Minimum Window Substring
* https://leetcode.com/problems/minimum-window-substring/description/

#### ['Bloomberg', 'Meta', 'Amazon', 'Apple', 'Uber']

```python
from collections import Counter

class Solution:
    def minWindow(self, s: str, t: str) -> str:
        need = Counter(t)            # freq of chars required
        missing = len(t)             # total chars still needed
        left = start = end = 0       # window pointers + result

        for right, ch in enumerate(s, 1):  # right is 1-based index
            if need[ch] > 0:         # if this char was needed
                missing -= 1
            need[ch] -= 1            # consume this char
            print(need)
            if missing == 0:         # window has all chars
                while left < right and need[s[left]] < 0:
                    need[s[left]] += 1
                    left += 1        # shrink from left
                
                print(need, s[left], right)

                if end == 0 or right - left < end - start:
                    start, end = left, right  # update best window

                print(start, end, right, left)
                need[s[left]] += 1   # remove leftmost char
                missing += 1
                left += 1            # move left to find next window

        return s[start:end]

Solution().minWindow('ADOBECODEBANC', 'ABC')
```
