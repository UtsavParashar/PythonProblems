```python
from collections import defaultdict, deque
# Definition for a binary tree node.
class TreeNode:
    """
        TC - O(n)
        SC - O(n)
    """
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def verticalTraversal(self, root):
        col_node_map = defaultdict(list)
        min_col_index = 0
        max_col_index = 0

        index_node_queue = deque([(0, root)])

        while index_node_queue:
            index, node = index_node_queue.popleft()
            col_node_map[index].append(node.val)

            if node.left:
                left_index = index-1
                index_node_queue.append((left_index, node.left))
                min_col_index = min(min_col_index, left_index) # remember this condition
            if node.right:
                right_index = index+1
                index_node_queue.append((right_index, node.right))
                max_col_index = max(max_col_index, right_index)
        
        return [col_node_map[col_index] for col_index in range(min_col_index, max_col_index)]
        
```

```python
from collections import defaultdict, deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def verticalTraversal(self, root):
        if not root:
            return []
        
        col_val_map = defaultdict(list)
        min_index = 0
        max_index = 0

        queue = deque([(0, root)])
        while queue:
            col, node = queue.popleft()
            col_val_map[col].append(node.val)

            if node.left:
                queue.append((col-1, node.left))
                min_index = min(min_index, col - 1)
            if node.right:
                queue.append((col+1, node.right))
                max_index = max(max_index, col + 1)
        
        return [col_val_map[index] for index in range(min_index, max_index+1)]
    

root = TreeNode(3)
root.left = TreeNode(9)
root.right = TreeNode(20)
root.right.left = TreeNode(15)
root.right.right = TreeNode(7)
Solution().verticalTraversal(root)
```

```python
from collections import defaultdict, deque

def kill_process(pid_, ppid_, kill):
    parent_child_graph = defaultdict(list)

    for pid, ppid in zip(pid_, ppid_):
        parent_child_graph[ppid].append(pid)

    
    visited = {kill}
    queue = deque([kill])
    killed_processes = []

    while queue:
        killed = queue.popleft()
        killed_processes.append(killed)

        for nei in parent_child_graph[killed]:
            if nei not in visited:
                visited.add(nei)
                queue.append(nei)

    return killed_processes



kill_process([1,3,10,5, 20], [3,0,5,3, 10], 5)

# 3 -> 1, 5
# 0 -> 3
# 5 -> 10

# 1 -> 3
# 3 -> 0
# 10 -> 5
# 5 -> 3
```
