#### 987. Vertical Order Traversal of a Binary Tree
* https://leetcode.com/problems/vertical-order-traversal-of-a-binary-tree/description/

```python
from collections import defaultdict

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def verticalTraversal(self, root):
        """
        ### DFS + Global Sort + Dict SOl - Preferred
        Two approaches - Same Time Complexity - O(n log n) as sorting is dominant
        Space Complexity - O(n)
        
        1. Extend the BFS + Dict Solution of LC 314 and add row to the storage and then sort based on column, then row and then value - requires too many sort methods and tedious col, row, val management

        2. DFS + Global Sort +  Dict Solution - Preferred as sort is global
        """

        nodes_list = [] # (col, row, vals)

        def traverse_tree_dfs(col, row, node):
            if not node:
                return

            nodes_list.append((col, row, node.val))

            traverse_tree_dfs(col-1, row+1, node.left)
            traverse_tree_dfs(col+1, row+1, node.right)

        traverse_tree_dfs(0, 0, root)

        # global sort 
        nodes_list.sort()
        
        col_val_map = defaultdict(list)
        for col, row, val in nodes_list:
            col_val_map[col].append(val)

        # all 3 are correct
        # return [col_val_map[col] for col in col_val_map.keys()]
        # return [col_val_map[col] for col in sorted(col_val_map)]
        return [val for col, val in col_val_map.items()]
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def verticalTraversal(self, root: Optional[TreeNode]) -> List[List[int]]:
        """
        Two approaches - Same Time Complexity - O(n log n) as sorting is dominant
        Space Complexity - O(n)
        
        1. Extend the BFS + Dict Solution of LC 314 and add row to the storage and then sort based on column, then row and then value - requires too many sort methods and tedious col, row, val management

        2. DFS + Global Sort +  Dict Solution - Preferred as sort is global
        """

        column_rol_val_map = defaultdict(list) # col -> (row, val)

        queue = deque([(0, 0, root)]) # (col, row, node)

        while queue:
            col, row, node = queue.popleft()

            column_rol_val_map[col].append((row, node.val))

            # row will always increase as we go down the tree
            if node.left:
                queue.append((col-1, row+1, node.left))
            if node.right:
                queue.append((col+1, row+1, node.right))

        res = []
        for col in sorted(column_rol_val_map): # sorted(d) return sorted keys
            # sort rows, vals for each key col
            col_vals = sorted(column_rol_val_map[col], key=lambda x: (x[0], x[1]))
            res.append([val for row, val in col_vals])

        return res
```
