# 834. Sum of Distances in Tree

* https://leetcode.com/problems/sum-of-distances-in-tree/description/

```python
class Solution:
    def sumOfDistancesInTree(self, n: int, edges: List[List[int]]) -> List[int]:
        dis = [0] * n
        child = [1] * n

        dd = defaultdict(list)

        for s, d in edges:
            dd[s].append(d)
            dd[d].append(s)

        # Calculate:
        # 1. child[node] -> size of subtree rooted at node
        # 2. dis[0] -> sum of distances from root (0) to all nodes
        def post_dfs(c, p):
            for nei in dd[c]:
                if nei == p:
                    continue

                post_dfs(nei, c)

                child[c] += child[nei]
                dis[c] += dis[nei] + child[nei]

        # Re-root DP:
        # dis[nei] = dis[c] + n - 2 * child[nei]
        def pre_dfs(c, p):
            for nei in dd[c]:
                if nei == p:
                    continue

                dis[nei] = dis[c] + n - 2 * child[nei]

                pre_dfs(nei, c)

        post_dfs(0, -1)
        pre_dfs(0, -1)

        return dis
```
