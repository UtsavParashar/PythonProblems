##### 1976. Number of Ways to Arrive at Destination

* https://leetcode.com/problems/number-of-ways-to-arrive-at-destination/description/

```python
from typing import List
from collections import defaultdict
import heapq

class Solution:
    def countPaths(self, n: int, roads: List[List[int]]) -> int:
        MOD = 1000_000_007

        dd = defaultdict(list)

        for s, d, w in roads:
            dd[s].append((d, w))
            dd[d].append((s, w))
        
        sd = [float('inf')]*(n) # shortest distance
        sd[0] = 0

        pw = [1]*(n) # path ways

        mh = [(0, 0)] # distance, node # min heap

        while mh:
            cd, cn = heapq.heappop(mh) # curr_dist, curr_node
            if cd > sd[cn]: 
                continue
            
            for nei, wt in dd[cn]:
                can_dis = cd + wt # candidate_distance
                if can_dis < sd[nei]:
                    sd[nei] = can_dis
                    pw[nei] = pw[cn]
                    heapq.heappush(mh, (can_dis, nei))
                elif can_dis == sd[nei]:
                    pw[nei] = (pw[cn] + pw[nei]) % MOD
        
        return pw[n-1]
```
