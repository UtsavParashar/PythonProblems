# PROBLEM
Follow-up:
What if you need to find the shortest string after removal?

Example:

Input: "aaabbbacd"
Output: "cd"
Explanation:
1. Remove 3 'b': "aaabbbacd" => "aaaacd"
2. Remove 4 'a': "aaaacd" => "cd"

```python
# This can be solved using dp as the follow is a State Search Problem, its no more a Greedy + Stack problem
# Memo/ Stack + BT solution would work here better as we don't know the states before hand
# We try to reduce string by skipping the chars using pointers(sliding) which occur more than k in str/substr and they check the SHORTEST string
# Tabulation might not be ideal here as we don't have predefined states

from functools import lru_cache

class Solution:
    def shortestAfterRemoval(self, s: str, k: int) -> str:
        """
        Returns the shortest possible string after removing
        any adjacent groups of size >= k in any order.

        Uses DFS + memoization to explore all valid collapse sequences.

        Time Complexity: Exponential in worst case, but heavily pruned by memo
        Space Complexity: O(states)

        """
        
        @lru_cache(None)
        def dfs(current):
            shortest = current
            n = len(shortest)
            i = 0
            while i < n:
                j = i
                while j < n and current[i] == current[j]:
                    j += 1
                
                # length of the run
                run_len = j - i
                if run_len >= k:
                    # If there is a substring with more than k same chars, let's remove it from next state
                    next_state = current[:i] + current[j:]
                    candidate = dfs(next_state)

                    if len(candidate) < len(shortest):
                        shortest = candidate
                
                i = j # this will help i and j pointer move forward in case of run_len less than k
            
            return shortest

        return dfs(s)

Solution().shortestAfterRemoval('aaabbbacd', 3)
```

```python
class Solution:
    def shortestAfterRemoval(self, s: str, k: int) -> str:
        """
        Returns the shortest possible string after removing
        any adjacent groups of size >= k in any order.

        DFS + manual memoization

        Time Complexity:
            Exponential worst case, but pruned by memoization

        Space Complexity:
            O(number_of_unique_states)
        """

        memo = {}

        def dfs(current: str) -> str:
            # Memo lookup
            if current in memo:
                return memo[current]

            shortest = current
            n = len(current)

            i = 0
            while i < n:
                j = i

                # find run of same characters
                while j < n and current[i] == current[j]:
                    j += 1

                run_len = j - i

                if run_len >= k:
                    next_state = current[:i] + current[j:]
                    candidate = dfs(next_state)

                    if len(candidate) < len(shortest):
                        shortest = candidate

                i = j

            memo[current] = shortest
            return shortest

        return dfs(s)


print(Solution().shortestAfterRemoval('aaabbbacd', 3))
```

```python
# This solution pass GFG - https://www.geeksforgeeks.org/problems/restrictive-candy-crush--141631/1

class Solution:
    def shortestAfterRemoval(self, s: str, k: int) -> str:
        memo = {}

        def crush(string: str) -> str:
            """
            Fully reduce the string using stack compression.
            """
            stack = []

            for ch in string:
                if stack and stack[-1][0] == ch:
                    stack[-1][1] += 1
                else:
                    stack.append([ch, 1])

                if stack[-1][1] >= k:
                    stack.pop()

            return "".join(c * cnt for c, cnt in stack)

        def dfs(current: str) -> str:
            current = crush(current)

            if current in memo:
                return memo[current]

            best = current
            n = len(current)

            i = 0
            while i < n:
                j = i
                while j < n and current[i] == current[j]:
                    j += 1

                run_len = j - i

                if run_len >= k:
                    next_state = current[:i] + current[j:]
                    candidate = dfs(next_state)

                    if len(candidate) < len(best):
                        best = candidate

                i = j

            memo[current] = best
            return best

        return dfs(s)
```
