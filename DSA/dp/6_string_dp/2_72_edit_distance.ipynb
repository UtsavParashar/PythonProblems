### 72. Edit Distance
* https://leetcode.com/problems/edit-distance/

### ['Bloomberg', 'DE Shaw', 'Meta', 'Amazon', 'Apple']

```python
# Same as prev quest LCS - 1143 and Maximal Square - 221
# Space Optimized solution of previous solution
# here only two rows of space is used at a time
# TC - O(m*n), SC - O(n)

class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        # very similar to Longest common subsequence
        # Edge cases
        # Both strings are empty - matched
        # word 1 is empty - all insert == len(word2)
        # word 2 is empty - all delete == len(word1)

        # 
        # Logic - if w1[i] == w2[j] then grid[i][j] = grid[i-1][j-1] as no operation is require
        # if w1[i] != w2[j] - choose delete, insert or replace
        # for insert = w1[i] would remain at the same place, w2 - j will increment
        # for delete = w1 i would increment, j will remain tthe same
        # for replace = w1 i would increment and w32 j will also increment
        # from  all 3 cases we would choose min at each cell and increment operation by 1

        if len(word1) < len(word2):
            word1, word2 = word2, word1

        m, n = len(word1), len(word2)

        prev = [0] + [i+1 for i in range(n)]
        for i, x in enumerate(word1, 1):
            curr = [i] + [0]*n
            for j, y in enumerate(word2, 1):
                if x == y:
                    curr[j] = prev[j-1]
                else:
                    curr[j] = 1 + min(
                        curr[j-1],
                        prev[j-1],
                        prev[j]
                    )
            prev = curr

        return prev[-1]
```

```python
# Great Explanation - https://www.youtube.com/watch?v=XYi2-LPrwm4
# TC - O(m*n), SC - O(m*n)
# Intuitive solution with more space

class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        m, n = len(word1), len(word2)
        dp = [[0]*(n+1) for _ in range(m+1)]

        # Base Case
        for i in range(m+1):
            dp[i][-1] = m-i # last column to be filled with len(word1)
        
        for j in range(n+1):
            dp[-1][j] = n-j # last row to be filled with len(word2)

        for i in range(m-1, -1, -1):
            for j in range(n-1, -1, -1):
                if word1[i] == word2[j]:
                    dp[i][j] = dp[i+1][j+1]
                else:
                    dp[i][j] = 1 + min(
                        dp[i+1][j], # delete
                        dp[i][j+1],  # insert
                        dp[i+1][j+1] # replace
                    )
        
        return dp[0][0]

Solution().minDistance('horse','ros')
```

```python
# Space Optimized solution of previous solution
# here only two rows of space is used at a time
# TC - O(m*n), SC - O(n)

class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        m, n = len(word1), len(word2)
        next_row = [n - j for j in range(n+1)]

        for i in range(m-1, -1, -1):
            curr_row = [0]*(n+1)
            curr_row[n] = m - i
            for j in range(n-1, -1, -1):
                if word1[i] == word2[j]:
                    curr_row[j] = next_row[j+1]
                else:
                    curr_row[j] = 1 + min(
                        next_row[j], # delete
                        curr_row[j+1], # insert
                        next_row[j+1] # replace
                    )
            next_row = curr_row
        
        return next_row[0]



Solution().minDistance('horse','ros')
```

🧠 Understanding the Recurrence:

curr_row[j] = 1 + min(
    next_row[j],     # delete
    curr_row[j + 1], # insert
    next_row[j + 1]  # replace
)
Each of these corresponds to an edit operation you can make at position i, j:

1️⃣ next_row[j] → Delete word1[i]
You delete word1[i], so now you’re left with word1[i+1:] and still matching it to word2[j:].

So: dp[i][j] = 1 + dp[i+1][j]

Which is: 1 + next_row[j] (because next_row = dp[i+1])

2️⃣ curr_row[j + 1] → Insert word2[j]
You insert word2[j] into word1, so the new word1 becomes word2[j] + word1[i:].

After inserting, you still need to match word1[i:] with word2[j+1:].

So: dp[i][j] = 1 + dp[i][j+1]

Which is: 1 + curr_row[j + 1] (because we’re currently building dp[i][...])

3️⃣ next_row[j + 1] → Replace word1[i] with word2[j]
You replace word1[i] with word2[j], then move to the next characters in both strings.

So: dp[i][j] = 1 + dp[i+1][j+1]

Which is: 1 + next_row[j + 1]

```python
import random


def estimate_pi(num_samples: int) -> float:
    """
    Estimate pi using Monte Carlo simulation.

    We sample random points uniformly inside
    the square [-1,1] x [-1,1].

    Fraction of points inside the unit circle
    approximates pi / 4.
    """

    inside_circle_count = 0

    for _ in range(num_samples):
        x = random.uniform(-1.0, 1.0)
        y = random.uniform(-1.0, 1.0)

        # Check whether point lies inside unit circle
        if x * x + y * y <= 1.0:
            inside_circle_count += 1

    return 4.0 * inside_circle_count / num_samples


# Example usage
estimated_pi = estimate_pi(1_000_000)

print(f"Estimated pi: {estimated_pi}")
```
