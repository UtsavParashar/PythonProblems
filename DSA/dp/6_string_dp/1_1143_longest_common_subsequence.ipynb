#### 1143. Longest Common Subsequence

* https://leetcode.com/problems/longest-common-subsequence/description/

* Prerequisite for https://leetcode.com/problems/edit-distance


```python
# Space Optimized solution
# TC - O(m*n)
# SC - O(n) 
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        if not text1 or not text2:
            return 0

        # For memory optimization, as we want to keep the row/string with min chars in memory at a time
        if len(text2) > len(text1):
            text1, text2 = text2, text1

        m, n = len(text1), len(text2)
        prev = [0]*(n+1)

        for x in text1:
            curr = [0]*(n+1)
            for j, y in enumerate(text2, 1):
                if x == y:
                    # to avoid double counting of duplicates like adc and aba - 
                    # a being present twice in the second str
                    curr[j] = 1 + prev[j-1] 
                else:
                    curr[j] = max(curr[j-1], prev[j])
            prev = curr

        return prev[-1]
```

```python

def longestCommonSubsequence(text1, text2):
    dp = [[0]*(len(text2)+1) for _ in range(len(text1)+1)]

    for i in range(1, len(text1)+1):
        for j in range(1, len(text2)+1):
            if text1[i-1] == text2[j-1]: # If the chars match, then add 1 and go diagonal
                dp[i][j] = 1+ dp[i-1][j-1]
            else:
                dp[i][j] = max(dp[i][j-1], dp[i-1][j])  # If chars does not match, go to the max of right or down
    return dp[len(text1)][len(text2)] # Final count will be at the 0th location
```
