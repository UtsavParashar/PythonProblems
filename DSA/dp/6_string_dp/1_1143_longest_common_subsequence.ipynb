#### 1143. Longest Common Subsequence

* https://leetcode.com/problems/longest-common-subsequence/description/

* Prerequisite for https://leetcode.com/problems/edit-distance


```python
# This problem is same as 221. Maximal Square in terms of logic
# https://leetcode.com/problems/maximal-square/description/

from typing import List


class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        """
        Computes the length of the Longest Common Subsequence (LCS)
        between two strings using space-optimized Dynamic Programming.

        Time Complexity: O(m * n)
        Space Complexity: O(n)

        Args:
            text1 (str): First input string
            text2 (str): Second input string

        Returns:
            int: Length of LCS
        """

        # Ensure text2 is the smaller string for better space usage
        if len(text1) < len(text2):
            text1, text2 = text2, text1

        len1, len2 = len(text1), len(text2)

        # dp[j] represents LCS for:
        # text1[:i-1] and text2[:j]
        dp: List[int] = [0] * (len2 + 1)

        for i in range(1, len1 + 1):
            prev_diag = 0  # This stores dp[i-1][j-1]

            for j in range(1, len2 + 1):
                temp = dp[j]  # Store current dp[j] before overwriting

                if text1[i - 1] == text2[j - 1]:
                    dp[j] = 1 + prev_diag
                else:
                    dp[j] = max(dp[j], dp[j - 1])

                prev_diag = temp  # Move diagonal forward

        return dp[len2]
```

```python
#Credit - https://www.youtube.com/watch?v=Ua0GhsJSlWM&t=283s - good explanation

# Intuitive solution but more space is required
# Time Complexity: O(m * n)
# Space Complexity: O(m* n)

class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        m, n = len(text1), len(text2)

        dp = [[0]*(n+1) for _ in range(m+1)]

        for i in range(m-1, -1, -1):
            for j in range(n-1, -1, -1):
                if text1[i] == text2[j]:
                    dp[i][j] = 1 + dp[i+1][j+1] # if chars match, go diagonal and add 1
                else:
                    dp[i][j] = max(dp[i+1][j], dp[i][j+1]) # else max of right or down
        
        return dp[0][0]
```
