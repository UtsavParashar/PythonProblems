#### 70. Climbing Stairs

* https://leetcode.com/problems/climbing-stairs/description/

You are climbing a staircase. It takes n steps to reach the top.

Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

 

Example 1:

Input: n = 2
Output: 2
Explanation: There are two ways to climb to the top.
1. 1 step + 1 step
2. 2 steps
Example 2:

Input: n = 3
Output: 3
Explanation: There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step


#### ['J.P. Morgan', 'Goldman Sachs', 'Barclays', 'Bloomberg', 'Intuit', 'Meta', 'Amazon', 'Apple', 'Uber']

##### In short, solution is fibonacci series starting with 1
* For 1 step there is 1 way
* for 2 steps there are 2 ways
* for 3 steps there are 3 ways
* for 4 steos there are 5 ways

```python
# While solving the 1 DP problem
# Think of state, that can the current state be determined by previous states in a linear structure
# Think of recurrence relation and base case

class Solution:
    def climbStairs(self, n: int) -> int:
        if n < 2:
            return n
        
        prev, curr = 1, 2

        for _ in range(3, n+1):
            prev, curr = curr, prev + curr
        
        return curr
```

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        if n < 2:
            return n
        
        prev2 = 1
        prev1 = 2

        for _ in range(2, n):
            prev2, prev1 = prev1, prev2 + prev1
        
        return prev1
```

```python
### bruteforce - recursion
## TC - O(2^n), SP - O(n)

def brute_force(n):
    if n <= 1:
        return 1

    return brute_force(n-1) + brute_force(n-2)
```

```python
brute_force(5)
```

```python
### DP with memoization - Top Down approach
## TC - O(n), SP  - O(n)
def dp_memo(n):
    memo = {1: 1, 2: 2}
    def f(n):
        if n in memo:
            return memo[n]
        else:
            memo[n] = f(n-1) + f(n-2)
        return memo[n]
    
    return f(n)
        
```

```python
dp_memo(5)
```

```python
### DP with Tabulation -  Bottom up Approach
## TC - O(n), SP  - O(n)

def dp_tab(n):
    if n <= 0: return n
    tab = [1, 2]

    for i in range(2, n):
        tab.append(tab[i-1] + tab[i-2])

    return tab[-1]
```

```python
dp_tab(5)
```

```python
### Most Optimized in terms of space and time
### DP with Tabulation - Bottom up Approach
## TC - O(n), SP  - O(1)

def dp_tab_op(n):
    if n <= 1: return n
    prev, curr = 1, 2

    for _ in range(2, n):
        prev, curr = curr, prev+curr

    return curr
```

```python
dp_tab_op(6)
```

```python
import math
def currency_arbitrage(rates):
    ccys = set()

    for u, v, _ in rates:
        ccys.add(u)
        ccys.add(v)
    
    n = len(ccys)
    ccys_to_ids = {ccy: id for id, ccy in enumerate(ccys)}

    edges = []
    for u, v, rate in rates:
        u_id = ccys_to_ids[u]
        v_id = ccys_to_ids[v]
        log_rate = -math.log(rate)
        edges.append((u_id, v_id, log_rate))
    
    distance = [float('inf')]*n
    distance[0] = 0

    for _ in range(n-1):
        for u, v, wt in edges:
            if distance[u] + wt < distance[v]:
                distance[v] = distance[u] + wt
    
    for u, v, wt in edges:
            if distance[u] + wt < distance[v]:
                return True
    
    return False

currency_arbitrage([['USD', 'EUR', 0.9], ['EUR', 'GBP', 0.8], ['GBP', 'USD', 1.2]])
```
