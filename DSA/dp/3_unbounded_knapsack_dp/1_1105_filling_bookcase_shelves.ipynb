#### 1105. Filling Bookcase Shelves

* https://leetcode.com/problems/filling-bookcase-shelves/description/

#### BBG - IMP

```python
from typing import List
import time

class Solution:
    def minHeightShelves(self, books: List[List[int]], shelfWidth: int) -> int:
        n = len(books)
        
        def dfs(start_index):
            if start_index == n:
                return 0

            current_width = 0
            current_height = 0
            min_height = float('inf')

            for current_index in range(start_index, n):
                book_width, book_height = books[current_index]

                current_width += book_width

                if current_width > shelfWidth:
                    break

                current_height = max(current_height, book_height)

                print(f'ci - {current_index}, ch - {current_height}, cw - {current_width}')
                time.sleep(10)

                remaining_height = dfs(current_index + 1)

                print(f'remaining_height - {remaining_height}')
                time.sleep(10)

                min_height = min(min_height, current_height + remaining_height)
                print(f'min_height - {min_height}')
                time.sleep(10)
            
            return min_height


        return dfs(0)
    
Solution().minHeightShelves([[1,1],[2,3],[2,3]], shelfWidth = 4)
```
